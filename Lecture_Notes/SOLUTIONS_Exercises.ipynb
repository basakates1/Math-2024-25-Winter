<h1> Basic Operations on Matrices </h1>

<hr/>

**1. Matrix Addition and Subtraction**

**a) A + B:**

$$
\mathbf{A}+\mathbf{B} = 
\begin{pmatrix}
1+5 & 2+6 \\
3+7 & 4+8 
\end{pmatrix} =
\begin{pmatrix}
6 & 8 \\
10 & 12 
\end{pmatrix}
$$

**b) B - A:**

$$
\mathbf{B}-\mathbf{A} = 
\begin{pmatrix}
5-1 & 6-2 \\
7-3 & 8-4 
\end{pmatrix} =
\begin{pmatrix}
4 & 4 \\
4 & 4 
\end{pmatrix}
$$

**c) A + C:**

$$
\mathbf{A}+\mathbf{C} = 
\begin{pmatrix}
1-1 & 2+2 \\
3+3 & 4+0 
\end{pmatrix} =
\begin{pmatrix}
0 & 4 \\
6 & 4 
\end{pmatrix}
$$

**d) D + E:**

**Note:** Matrices D and E have different dimensions. We cannot add matrices with different dimensions. 
Therefore, D + E is undefined.

**2. Scalar Multiplication**

**a) (1/2)A:**

$$\frac{1}{2}\mathbf{A} = \frac{1}{2}
\begin{pmatrix}
1 & 2 \\
3 & 4 
\end{pmatrix} =
\begin{pmatrix}
\frac{1}{2} & 1 \\
\frac{3}{2} & 2 
\end{pmatrix}
$$

**b) 2B:**

$$2\mathbf{B} = 2
\begin{pmatrix}
5 & 6 \\
7 & 8 
\end{pmatrix} =
\begin{pmatrix}
10 & 12 \\
14 & 16 
\end{pmatrix}
$$

**c) -3C:**

$$-3\mathbf{C} = -3
\begin{pmatrix}
-1 & 2 \\
3 & 0 
\end{pmatrix} =
\begin{pmatrix}
3 & -6 \\
-9 & 0 
\end{pmatrix}
$$

**d) 4D:**

$$4\mathbf{D} = 4
\begin{pmatrix}
-1 & 2 & 3 \\
4 & 0 & 6 
\end{pmatrix} =
\begin{pmatrix}
-4 & 8 & 12 \\
16 & 0 & 24 
\end{pmatrix}
$$

**3. Matrix Multiplication**

**a) A * B:**

$$
\mathbf{A} \cdot \mathbf{B} = 
\begin{pmatrix}
1 & 2 \\
3 & 4 
\end{pmatrix}
\cdot
\begin{pmatrix}
5 & 6 \\
7 & 8 
\end{pmatrix} =
\begin{pmatrix}
19 & 22 \\
43 & 50 
\end{pmatrix}
$$

**b) B * A:**

$$
\mathbf{B} \cdot \mathbf{A} = 
\begin{pmatrix}
5 & 6 \\
7 & 8 
\end{pmatrix}
\cdot
\begin{pmatrix}
1 & 2 \\
3 & 4 
\end{pmatrix} =
\begin{pmatrix}
23 & 34 \\
31 & 46 
\end{pmatrix}
$$

**c) A * D:**

**Note:** The number of columns in A (2) does not match the number of rows in D (2). Therefore, the product A * D is undefined.

**d) D * E:**

$$
\mathbf{D} \cdot \mathbf{E} = 
\begin{pmatrix}
-1 & 2 & 3 \\
4 & 0 & 6 
\end{pmatrix}
\cdot
\begin{pmatrix}
1 & 2\\

\end{pmatrix} =
\begin{pmatrix}

50 & 68 
\end{pmatrix}

<h1> 2. Determinants 2x2 and 3x3 <h1>

<hr/>


### 1. 2×2 Matrices

Recall that for a 2×2 matrix

$$
\begin{pmatrix}
a & b\\
c & d
\end{pmatrix},
$$

its determinant is given by

$$
\det = ad - bc.
$$

**Matrix A**

$$
A = \begin{pmatrix}
2 & 3\\
1 & 4
\end{pmatrix}
$$

$$
\det(A) = 2 \cdot 4 - 3 \cdot 1 = 8 - 3 = 5.
$$

**Matrix B**

$$
B = \begin{pmatrix}
5 & 6\\
7 & 8
\end{pmatrix}
$$

$$
\det(B) = 5 \cdot 8 - 6 \cdot 7 = 40 - 42 = -2.
$$

**Matrix C**

$$
C = \begin{pmatrix}
-1 & 2\\
3 & 0
\end{pmatrix}
$$

$$
\det(C) = (-1) \cdot 0 - 2 \cdot 3 = 0 - 6 = -6.
$$

### 2. 3×3 Matrices

For a 3×3 matrix

$$
\begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{pmatrix},
$$

one common way to compute the determinant is by cofactor (Laplace) expansion along the first row:

$$
\det = a \begin{vmatrix} e & f \\ h & i \end{vmatrix} - b \begin{vmatrix} d & f \\ g & i \end{vmatrix} + c \begin{vmatrix} d & e \\ g & h \end{vmatrix}.
$$

**Matrix D**

$$
D = \begin{pmatrix}
1 & 0 & 2\\
-1 & 3 & 1\\
2 & 4 & -2
\end{pmatrix}
$$

Expand along the first row:

$$
\det(D) = 1 \cdot \begin{vmatrix} 3 & 1\\ 4 & -2 \end{vmatrix} - 0 \cdot \begin{vmatrix} -1 & 1\\ 2 & -2 \end{vmatrix} + 2 \cdot \begin{vmatrix} -1 & 3\\ 2 & 4 \end{vmatrix}.
$$

Minor for the entry 1 in the (1,1) position:

$$
\begin{vmatrix} 3 & 1\\ 4 & -2 \end{vmatrix} = 3 \cdot (-2) - 1 \cdot 4 = -6 - 4 = -10.
$$

Contribution: $$1 \times (-10) = -10$$.

The second term has a factor of 0, so it contributes 0.

Minor for the entry 2 in the (1,3) position:

$$
\begin{vmatrix} -1 & 3\\ 2 & 4 \end{vmatrix} = (-1)\cdot 4 - 3 \cdot 2 = -4 - 6 = -10.
$$

Contribution: $$2 \times (-10) = -20$$.

Therefore,

$$
\det(D) = -10 + 0 + (-20) = -30.
$$

**Matrix E**

$$
E = \begin{pmatrix}
3 & 1 & -1\\
0 & 2 & 4\\
5 & 3 & 2
\end{pmatrix}.
$$

Expand along the first row:

$$
\det(E) = 3 \cdot \begin{vmatrix} 2 & 4\\ 3 & 2 \end{vmatrix} - 1 \cdot \begin{vmatrix} 0 & 4\\ 5 & 2 \end{vmatrix} + (-1) \cdot \begin{vmatrix} 0 & 2\\ 5 & 3 \end{vmatrix}
$$

Minor for the entry 3 in the (1,1) position:

$$
\begin{vmatrix} 2 & 4\\ 3 & 2 \end{vmatrix} = 2 \cdot 2 - 4 \cdot 3 = 4 - 12 = -8.
$$

Contribution: $$3 \times (-8) = -24$$.

Minor for the entry 1 in the (1,2) position:

$$
\begin{vmatrix} 0 & 4\\ 5 & 2 \end{vmatrix} = 0 \cdot 2 - 4 \cdot 5 = 0 - 20 = -20.
$$

Contribution: $$= -1 \times (-20) = +20$$

Minor for the entry (-1) in the (1,3) position:

$$
\begin{vmatrix} 0 & 2\\ 5 & 3 \end{vmatrix} = 0 \cdot 3 - 2 \cdot 5 = 0 - 10 = -10
$$

Contribution: $$= (-1) \times (-10) = +10$$

Hence,

$$
\det(E) = (-24) + 20 + 10 = 6.
$$

**Matrix F**

$$
F = \begin{pmatrix}
2 & -3 & 1\\
1 & 4 & -2\\
1 & 5 & 3
\end{pmatrix}
$$

Expand along the first row:

$$
\det(F) = 2 \cdot \begin{vmatrix} 4 & -2\\ 5 & 3 \end{vmatrix} - (-3) \cdot \begin{vmatrix} 1 & -2\\ 1 & 3 \end{vmatrix} + 1 \cdot \begin{vmatrix} 1 & 4\\ 1 & 5 \end{vmatrix}.
$$

Minor for the entry 2 in the (1,1) position:

$$
\begin{vmatrix} 4 & -2\\ 5 & 3 \end{vmatrix} = 4 \cdot 3 - (-2) \cdot 5 = 12 + 10 = 22.
$$

Contribution: $$2 \times 22 = 44$$

Minor for the entry (-3) in the (1,2) position:

$$
\begin{vmatrix} 1 & -2\\ 1 & 3 \end{vmatrix} = 1 \cdot 3 - (-2) \cdot 1 = 3 + 2 = 5.
$$

Contribution $$= -(-3) \times 5 = +3 \times 5 = 15$$

Minor for the entry 1 in the (1,3) position:

$$
\begin{vmatrix} 1 & 4\\ 1 & 5 \end{vmatrix} = 1 \cdot 5 - 4 \cdot 1 = 5 - 4 = 1.
$$

Contribution: $$1 \times 1 = 1$$

Hence,

$$
\det(F) = 44 + 15 + 1 = 60.
$$

### Final Answers

**2×2 Matrices**

$$
\det(A) = 5, \quad \det(B) = -2, \quad \det(C) = -6.
$$

**3×3 Matrices**

$$
\det(D) = -30, \quad \det(E) = 6, \quad \det(F) = 60.
$$

<h1>3. Determinants using Laplace's Expansion</h1>

<hr/>

**1) Matrix A**

$$
A = \begin{pmatrix}
2 & 3 & 1 \\
1 & 4 & 0 \\
3 & 2 & 1
\end{pmatrix}.
$$

Expanding along row 1:

$$
\det(A) = 2\begin{vmatrix}4 & 0 \\ 2 & 1\end{vmatrix} - 3\begin{vmatrix}1 & 0 \\ 3 & 1\end{vmatrix} + 1\begin{vmatrix}1 & 4 \\ 3 & 2\end{vmatrix}.
$$

$$
\begin{vmatrix}4 & 0 \\ 2 & 1\end{vmatrix} = 4\cdot 1 - 0\cdot 2 = 4.
$$

$$
\begin{vmatrix}1 & 0 \\ 3 & 1\end{vmatrix} = 1\cdot 1 - 0\cdot 3 = 1.
$$

$$
\begin{vmatrix}1 & 4 \\ 3 & 2\end{vmatrix} = 1\cdot 2 - 4\cdot 3 = 2 - 12 = -10.
$$

Hence

$$
\det(A) = 2\cdot 4 - 3\cdot 1 + 1\cdot(-10) = 8 - 3 - 10 = -5.
$$

**2) Matrix B**

$$
B = \begin{pmatrix}
2 & 3 & 1 \\
1 & 4 & 0 \\
3 & 2 & 0
\end{pmatrix}.
$$

Expanding along row 1:

$$
\det(B) = 2\begin{vmatrix}4 & 0 \\ 2 & 0\end{vmatrix} - 3\begin{vmatrix}1 & 0 \\ 3 & 0\end{vmatrix} + 1\begin{vmatrix}1 & 4 \\ 3 & 2\end{vmatrix}.
$$

$$
\begin{vmatrix}4 & 0 \\ 2 & 0\end{vmatrix} = 4\cdot 0 - 0\cdot 2 = 0.
$$

$$
\begin{vmatrix}1 & 0 \\ 3 & 0\end{vmatrix} = 1\cdot 0 - 0\cdot 3 = 0.
$$

$$
\begin{vmatrix}1 & 4 \\ 3 & 2\end{vmatrix} = 1\cdot 2 - 4\cdot 3 = 2 - 12 = -10.
$$

Thus

$$
\det(B) = 2\cdot 0 - 3\cdot 0 + 1\cdot(-10) = -10.
$$

**3) Matrix C**

$$
C = \begin{pmatrix}
2 & 3 & 1 & 4 \\
1 & 0 & 0 & 6 \\
3 & 2 & 1 & 5 \\
2 & 1 & 4 & 0
\end{pmatrix}.
$$

Expanding along row 2:

$$
\det(C) = 1\cdot \mathrm{cofactor}(2,1) + 6\cdot \mathrm{cofactor}(2,4).
$$

$$
\mathrm{cofactor}(2,1) = (-1)^{2+1}\begin{vmatrix}3 & 1 & 4\\2 & 1 & 5\\1 & 4 & 0\end{vmatrix} = (-1)(-27) = 27.
$$

$$
\mathrm{cofactor}(2,4) = (-1)^{2+4}\begin{vmatrix}2 & 3 & 1\\3 & 2 & 1\\2 & 1 & 4\end{vmatrix} = (1)(-17) = -17.
$$

Thus

$$
\det(C) = 1\cdot 27 + 6\cdot (-17) = 27 - 102 = -75.
$$

**4) Matrix D**

$$
D = \begin{pmatrix}
2 & 3 & 1 & 4 & 5 \\
1 & 4 & 0 & 0 & 7 \\
3 & 0 & 0 & 0 & 0 \\
2 & 1 & 4 & 3 & 2 \\
1 & 2 & 3 & 4 & 5
\end{pmatrix}.
$$

Expanding along row 3:

$$
\det(D) = 3\cdot \mathrm{cofactor}(3,1).
$$

$$
\mathrm{cofactor}(3,1) = (-1)^{3+1}\begin{vmatrix}3 & 1 & 4 & 5\\4 & 0 & 0 & 7\\1 & 4 & 3 & 2\\2 & 3 & 4 & 5\end{vmatrix} = \begin{vmatrix}3 & 1 & 4 & 5\\4 & 0 & 0 & 7\\1 & 4 & 3 & 2\\2 & 3 & 4 & 5\end{vmatrix}.
$$

Expanding this 4x4 determinant along its second row:

$$
\det(M_{3,1}) = 4\cdot \mathrm{cofactor}(2,1) + 7\cdot \mathrm{cofactor}(2,4)
$$
$$
= 4\begin{vmatrix}1 & 4 & 5 \\ 4 & 3 & 2 \\ 3 & 4 & 5 \end{vmatrix} + 7\begin{vmatrix}3 & 1 & 4 \\ 1 & 4 & 3 \\ 2 & 3 & 4 \end{vmatrix}
$$
$$
= 4(14) + 7(3) = 56 + 21 = 77
$$

Finally,

$$
\det(D) = 3 \times 77 = 231.
$$

**Final Results**

$$
\det(A) = -5, \quad \det(B) = -10, \quad \det(C) = -75, \quad \det(D) = 231.
$$

<h1> 4. Determinants from the Gauss Method and Triangular Matrices </h1>

<hr/>

## Determinants using Row Operations

**1. Matrix A**

$$
A = \begin{pmatrix}
12 & 3 \\
-18 & -4
\end{pmatrix}.
$$

**Step 1: Use Row Operations to Get Upper Triangular Form**

We want a zero below the main diagonal in the position (2,1). Currently, that entry is -18. We can eliminate it by adding a suitable multiple of Row 1 to Row 2.

Let $R_2 \to R_2 + \frac{3}{2}R_1$.

Why $\frac{3}{2}$? Because $-18 + \frac{3}{2} \times 12 = -18 + 18 = 0$.

Performing this on each entry in Row 2:

First entry of new $R_2$: $-18 \;+\;\frac{3}{2}\times 12 \;=\;-18 \;+\;18 \;=\; 0$.

Second entry of new $R_2$: $-4 \;+\;\frac{3}{2}\times 3 \;=\;-4 \;+\;\frac{9}{2} \;=\;-\frac{8}{2} + \frac{9}{2} \;=\;\frac{1}{2}$.

Hence, after this row operation,

$$
\begin{pmatrix}
12 & 3 \\
-18 & -4
\end{pmatrix}
\;\xrightarrow{R_2 \to R_2+\frac{3}{2}R_1}\;
\begin{pmatrix}
12 & 3 \\
0 & \frac{1}{2}
\end{pmatrix}.
$$

This is now upper triangular.

**Step 2: Determinant from Diagonal**

Since we only added a multiple of one row to another (an operation that does not change the determinant), the determinant of $A$ is just the product of the diagonal of the resulting matrix:

$$
\det(A) \;=\; 12 \times \frac{1}{2} \;=\; 6.
$$

(You can also verify by the direct 2×2 determinant formula:
$\det(A) = 12 \cdot (-4)\;-\;3 \cdot (-18) = -48 + 54 = 6$.)

**2. Matrix B**

$$
B = \begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{pmatrix}.
$$

**Step 1: Eliminate Below the First Pivot**

We want zeros below the (1,1) entry (which is 1) in the first column:

Eliminate entry (2,1)=4 by using $R_2 \to R_2 - 4R_1$:

$R_2 - 4R_1 = \begin{pmatrix} 4 & 5 & 6 \end{pmatrix} - 4 \times \begin{pmatrix} 1 & 2 & 3 \end{pmatrix} = \begin{pmatrix} 4-4 & 5-8 & 6-12 \end{pmatrix} = \begin{pmatrix} 0 & -3 & -6 \end{pmatrix}$.

Eliminate entry (3,1)=7 by using $R_3 \to R_3 - 7R_1$:

$R_3 - 7R_1 = \begin{pmatrix} 7 & 8 & 9 \end{pmatrix} - 7 \times \begin{pmatrix} 1 & 2 & 3 \end{pmatrix} = \begin{pmatrix} 7-7 & 8-14 & 9-21 \end{pmatrix} = \begin{pmatrix} 0 & -6 & -12 \end{pmatrix}$.

So now $B$ becomes:

$$
\begin{pmatrix}
1 & 2 & 3 \\
0 & -3 & -6 \\
0 & -6 & -12
\end{pmatrix}.
$$

**Step 2: Eliminate Below the Second Pivot**

Next, focus on the second pivot, the entry (2,2)=-3. We want a zero in position (3,2). Perform $R_3 \to R_3 - 2R_2$:

$R_3 - 2R_2 = \begin{pmatrix} 0 & -6 & -12 \end{pmatrix} - 2 \times \begin{pmatrix} 0 & -3 & -6 \end{pmatrix} = \begin{pmatrix} 0 & -6 & -12 \end{pmatrix} - \begin{pmatrix} 0 & -6 & -12 \end{pmatrix} = \begin{pmatrix} 0 & 0 & 0 \end{pmatrix}$.

Thus $B$ is now in upper‐triangular form:

$$
\begin{pmatrix}
1 & 2 & 3 \\
0 & -3 & -6 \\
0 & 0 & 0
\end{pmatrix}.
$$

**Step 3: Determinant from Diagonal**

The diagonal entries are 1, -3, 0. The product is

$$
\det(B) \;=\; 1 \;\times\;(-3)\;\times\;0 \;=\; 0.
$$

Because one of the diagonal entries is zero, $\det(B) = 0$. (In fact, $B$ is well-known to be singular because its rows are linearly dependent.)

**Final Answers**

$$
\det(A) = 6.
$$

$$
\det(B) = 0.
$$

<h3><b>SOLUTION 1</b></h3>

We want to find the inverse of the 3×3 matrix

$A = \begin{pmatrix}
2 & 0 & 1 \\
0 & 1 & 0 \\
1 & 2 & 0
\end{pmatrix}.$

Recall that the inverse $A^{-1}$ satisfies

$A \, A^{-1} = I,$

where $I$ is the 3×3 identity matrix.

1. Compute $\det(A)$

Using the cofactor expansion (along the first row, for instance):

$\det(A) = 2 \cdot \det\begin{pmatrix} 1 & 0 \\ 2 & 0 \end{pmatrix} - 0 \cdot \dots + 1 \cdot \det\begin{pmatrix} 0 & 1 \\ 1 & 2 \end{pmatrix}.$

$\det\begin{pmatrix} 1 & 0 \\ 2 & 0 \end{pmatrix} = 1\cdot0 - 0\cdot2 = 0.$

Thus the first term $= 2 \times 0 = 0.$

The middle term is multiplied by 0, so it is 0.

$\det\begin{pmatrix} 0 & 1 \\ 1 & 2 \end{pmatrix} = 0\cdot2 - 1\cdot1 = -1.$

Hence,

$\det(A) = 0 + 0 + 1 \cdot (-1) = -1.$

Since $\det(A) = -1 \neq 0$, the matrix $A$ is invertible.

2. Find the adjugate $\mathrm{adj}(A)$

The formula for the inverse is:

$A^{-1} = \frac{1}{\det(A)} \,\mathrm{adj}(A).$

*2.1. Cofactor matrix*

Minor $M_{1,1}$: remove row 1, col 1
$\begin{vmatrix} 1 & 0 \\ 2 & 0 \end{vmatrix} = 1\cdot 0 - 0\cdot 2 = 0$
$C_{1,1} = (+1)\cdot 0 = 0.$

Minor $M_{1,2}$: remove row 1, col 2
$\begin{vmatrix} 0 & 0 \\ 1 & 0 \end{vmatrix} = 0\cdot 0 - 0\cdot 1 = 0$
$C_{1,2} = (-1)\cdot 0 = 0.$

Minor $M_{1,3}$: remove row 1, col 3
$\begin{vmatrix} 0 & 1 \\ 1 & 2 \end{vmatrix} = 0\cdot 2 - 1\cdot 1 = -1.$
$C_{1,3} = (+1)\cdot (-1) = -1.$

Minor $M_{2,1}$: remove row 2, col 1
$\begin{vmatrix} 0 & 1 \\ 2 & 0 \end{vmatrix} = 0\cdot 0 - 1\cdot 2 = -2.$
$C_{2,1} = (-1)\cdot (-2) = +2.$

Minor $M_{2,2}$: remove row 2, col 2
$\begin{vmatrix} 2 & 1 \\ 1 & 0 \end{vmatrix} = 2\cdot 0 - 1\cdot 1 = -1.$
$C_{2,2} = (+1)\cdot (-1) = -1.$

Minor $M_{2,3}$: remove row 2, col 3
$\begin{vmatrix} 2 & 0 \\ 1 & 2 \end{vmatrix} = 2\cdot 2 - 0\cdot 1 = 4.$
$C_{2,3} = (-1)\cdot 4 = -4.$

Minor $M_{3,1}$: remove row 3, col 1
$\begin{vmatrix} 0 & 1 \\ 1 & 0 \end{vmatrix} = 0\cdot 0 - 1\cdot 1 = -1.$
$C_{3,1} = (+1)\cdot (-1) = -1.$

Minor $M_{3,2}$: remove row 3, col 2
$\begin{vmatrix} 2 & 1 \\ 0 & 0 \end{vmatrix} = 2\cdot 0 - 1\cdot 0 = 0.$
$C_{3,2} = (-1)\cdot 0 = 0.$

Minor $M_{3,3}$: remove row 3, col 3
$\begin{vmatrix} 2 & 0 \\ 0 & 1 \end{vmatrix} = 2\cdot 1 - 0\cdot 0 = 2.$
$C_{3,3} = (+1)\cdot 2 = 2.$

Hence, the cofactor matrix $C$ is

$C = \begin{pmatrix}
0 & 0 & -1 \\
2 & -1 & -4 \\
-1 & 0 & 2
\end{pmatrix}.$

2.2. Adjugate matrix

$\mathrm{adj}(A) = C^\mathsf{T} = \begin{pmatrix}
0 & 2 & -1 \\
0 & -1 & 0 \\
-1 & -4 & 2
\end{pmatrix}.$

3. Compute $A^{-1}$

Since $\det(A) = -1$, we have

$A^{-1} = \frac{1}{\det(A)} \,\mathrm{adj}(A) = \frac{1}{-1} \,\mathrm{adj}(A) = -\,\mathrm{adj}(A).$

Multiplying each entry of $\mathrm{adj}(A)$ by $-1$ gives

$A^{-1} = \begin{pmatrix}
0 & -2 & 1 \\
0 & 1 & 0 \\
1 & 4 & -2
\end{pmatrix}.$

4. Verification

To verify, we check $A \cdot A^{-1} = I$. Let’s multiply:

$A \cdot A^{-1} = \begin{pmatrix} 2 & 0 & 1 \\ 0 & 1 & 0 \\ 1 & 2 & 0 \end{pmatrix} \begin{pmatrix} 0 & -2 & 1 \\ 0 & 1 & 0 \\ 1 & 4 & -2 \end{pmatrix}.$

Row 1 × Column 1: $2\cdot0 + 0\cdot0 + 1\cdot1 = 1$

Row 1 × Column 2: $2\cdot(-2) + 0\cdot1 + 1\cdot4 = -4 + 4 = 0$

Row 1 × Column 3: $2\cdot1 + 0\cdot0 + 1\cdot(-2) = 2 - 2 = 0$

Row 2 × Column 1: $0\cdot0 + 1\cdot0 + 0\cdot1 = 0$

Row 2 × Column 2: $0\cdot(-2) + 1\cdot1 + 0\cdot4 = 1$

Row 2 × Column 3: $0\cdot1 + 1\cdot0 + 0\cdot(-2) = 0$

Row 3 × Column 1: $1\cdot0 + 2\cdot0 + 0\cdot1 = 0$

Row 3 × Column 2: $1\cdot(-2) + 2\cdot1 + 0\cdot4 = -2 + 2 = 0$

Row 3 × Column 3: $1\cdot1 + 2\cdot0 + 0\cdot(-2) = 1$

Hence,

$A \,A^{-1} = \begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{pmatrix} = I.$

This confirms the correctness of $A^{-1}$.



<h3><b> SOLUTION 2 </b></h3>

To determine the rank of the 3×3 matrix

$B = \begin{pmatrix}
4 & -3 & 7 \\
-1 & 6 & 3 \\
2 & 9 & 1
\end{pmatrix},$

we can check its determinant. If the determinant is nonzero, the matrix is invertible and hence has rank 3.

Step-by-step determinant calculation

Using cofactor expansion along the first row:

$\det(B) = 4 \cdot \begin{vmatrix}
6 & 3 \\
9 & 1
\end{vmatrix} - (-3)\cdot \begin{vmatrix}
-1 & 3 \\
2 & 1
\end{vmatrix} + 7 \cdot \begin{vmatrix}
-1 & 6 \\
2 & 9
\end{vmatrix}.$

First minor (remove row 1, column 1):

$\begin{vmatrix}
6 & 3 \\
9 & 1
\end{vmatrix} = 6 \cdot 1 - 3 \cdot 9 = 6 - 27 = -21.$

Contribution to determinant: $4 \times (-21) = -84.$

Second minor (remove row 1, column 2):

$\begin{vmatrix}
-1 & 3 \\
2 & 1
\end{vmatrix} = (-1) \cdot 1 - 3 \cdot 2 = -1 - 6 = -7.$

Contribution to determinant: $-(-3) \times (-7) = -21.$

Third minor (remove row 1, column 3):

$\begin{vmatrix}
-1 & 6 \\
2 & 9
\end{vmatrix} = (-1) \cdot 9 - 6 \cdot 2 = -9 - 12 = -21.$

Contribution to determinant: $+7 \times (-21) = -147.$

Putting these together:

$\det(B) = -84 - 21 - 147 = -252.$

Since $\det(B) = -252 \neq 0$, the matrix $B$ is invertible. Therefore,

$\text{rank}(B) = 3.$

<h2><b> 6. Inverse of a Matrix using the Gauss Method </b></h2>

<hr/>

1. Inverse of:

$A = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}$.

Method Overview (2×2 formula)

For a 2×2 matrix $\begin{pmatrix} a & b\\ c & d \end{pmatrix}$, the inverse (if it exists) is given by

$\begin{pmatrix} a & b\\ c & d \end{pmatrix}^{-1} = \frac{1}{ad - bc} \begin{pmatrix} d & -b\\ -c & a \end{pmatrix}$.

Step 1: Compute the determinant

$\det(A) = (1)(4) - (2)(3) = 4 - 6 = -2$.
Since $\det(A) = -2 \ne 0$, $A$ is invertible.

Step 2: Apply the formula

$A^{-1} = \frac{1}{-2} \begin{pmatrix} 4 & -2\\ -3 & 1 \end{pmatrix} = \begin{pmatrix} \frac{4}{-2} & \frac{-2}{-2}\\ \frac{-3}{-2} & \frac{1}{-2} \end{pmatrix} = \begin{pmatrix} -2 & 1\\ \frac{3}{2} & -\frac{1}{2} \end{pmatrix}$.

Hence,

$A^{-1} = \begin{pmatrix} -2 & 1\\ \frac{3}{2} & -\frac{1}{2} \end{pmatrix}$.

2. Inverse of
$B = \begin{pmatrix} 1 & 2 & 3\\ 4 & 5 & 1\\ 2 & 3 & 2 \end{pmatrix}$.

Method Overview
For a 3×3 matrix $B$, one systematic approach is:

*   Compute $\det(B)$.
*   Compute the cofactor matrix.
*   Transpose the cofactor matrix.
*   Multiply by $1/\det(B)$.

Step 1: Compute $\det(B)$

Using the first row for expansion:

$\det(B) = 1 \begin{vmatrix} 5 & 1 \\ 3 & 2 \end{vmatrix} - 2 \begin{vmatrix} 4 & 1 \\ 2 & 2 \end{vmatrix} + 3 \begin{vmatrix} 4 & 5 \\ 2 & 3 \end{vmatrix}$.

Minor for the first term: $(5)(2) - (1)(3) = 10 - 3 = 7$.

Minor for the second term: $(4)(2) - (1)(2) = 8 - 2 = 6$.

Minor for the third term: $(4)(3) - (5)(2) = 12 - 10 = 2$.

Putting it all together:

$\det(B) = 1\cdot (7) - 2\cdot (6) + 3\cdot (2) = 7 - 12 + 6 = 1$.

Since $\det(B) = 1$, $B$ is invertible.

Step 2: Cofactor matrix of $B$

Let $B = (b_{ij})$. The $(i,j)$-minor $M_{ij}$ is the determinant of the (2×2) submatrix obtained by removing row $i$ and column $j$. The cofactor is $C_{ij} = (-1)^{i+j} M_{ij}$.

$C_{11}$: $\begin{vmatrix} 5 & 1 \\ 3 & 2 \end{vmatrix} = 5\cdot 2 - 1\cdot 3 = 7$. Sign $(+1)$. So $C_{11}= +7$.

$C_{12}$: $\begin{vmatrix} 4 & 1 \\ 2 & 2 \end{vmatrix} =4\cdot 2 - 1\cdot 2=6$. Sign $(-1)$. So $C_{12}= -6$.

$C_{13}$: $\begin{vmatrix} 4 & 5 \\ 2 & 3 \end{vmatrix} =4\cdot 3 - 5\cdot 2=2$. Sign $(+1)$. So $C_{13}= +2$.

$C_{21}$: $\begin{vmatrix} 2 & 3 \\ 3 & 2 \end{vmatrix} =2\cdot 2 - 3\cdot 3=-5$. Sign $(-1)$. So $C_{21}= +5$.

$C_{22}$: $\begin{vmatrix} 1 & 3 \\ 2 & 2 \end{vmatrix} =1\cdot 2 - 3\cdot 2=-4$. Sign $(+1)$. So $C_{22}= -4$.

$C_{23}$: $\begin{vmatrix} 1 & 2 \\ 2 & 3 \end{vmatrix} =1\cdot 3 - 2\cdot 2=-1$. Sign $(-1)$. So $C_{23}= +1$.

$C_{31}$: $\begin{vmatrix} 2 & 3 \\ 5 & 1 \end{vmatrix} =2\cdot 1 - 3\cdot 5=-13$. Sign $(+1)$. So $C_{31}= -13$.

$C_{32}$: $\begin{vmatrix} 1 & 3 \\ 4 & 1 \end{vmatrix} =1\cdot 1 - 3\cdot 4=-11$. Sign $(-1)$. So $C_{32}= +11$.

$C_{33}$: $\begin{vmatrix} 1 & 2 \\ 4 & 5 \end{vmatrix} =1\cdot 5 - 2\cdot 4=-3$. Sign $(+1)$. So $C_{33}= -3$.

$C(B) = \begin{pmatrix} 7 & -6 & 2 \\ 5 & -4 & 1 \\ -13 & 11 & -3 \end{pmatrix}$.

Step 3: Transpose the cofactor matrix (adjugate)

$\text{adj}(B) = C(B)^\mathsf{T} = \begin{pmatrix} 7 & 5 & -13\\ -6 & -4 & 11\\ 2 & 1 & -3 \end{pmatrix}$.

Step 4: Multiply by $1/\det(B)$

Since $\det(B)=1$, we have

$B^{-1} = \text{adj}(B)$.

Hence,

$B^{-1} = \begin{pmatrix} 7 & 5 & -13\\ -6 & -4 & 11\\ 2 & 1 & -3 \end{pmatrix}$.

3. Inverse of:

$C = \begin{pmatrix} 0 & 0 & 1\\ 0 & 1 & 0\\ 1 & 0 & 0 \end{pmatrix}$.

Observe that $C$ is a permutation matrix.

$C^2 = I$.

Hence $C^{-1} = C$.

$C\cdot C = \begin{pmatrix} 0 & 0 & 1\\ 0 & 1 & 0\\ 1 & 0 & 0 \end{pmatrix}\begin{pmatrix} 0 & 0 & 1\\ 0 & 1 & 0\\ 1 & 0 & 0 \end{pmatrix} = I$.

Therefore,
$C^{-1} = \begin{pmatrix} 0 & 0 & 1\\ 0 & 1 & 0\\ 1 & 0 & 0 \end{pmatrix}$.

Final Answers

$A^{-1} = \begin{pmatrix} -2 & 1\\ \frac{3}{2} & -\frac{1}{2} \end{pmatrix}$.

$B^{-1} = \begin{pmatrix} 7 & 5 & -13\\ -6 & -4 & 11\\ 2 & 1 & -3 \end{pmatrix}$.

$$
\begin{cases}
3x - 2y = 5 \\
2x + 3y = 7
\end{cases}
$$

**Step 1. Eliminate one variable (e.g.\y).**

Multiply the first equation by 3 and the second by 2:

$$
\begin{aligned}
(3x - 2y) \cdot 3 &: 9x - 6y = 15 \\
(2x + 3y) \cdot 2 &: 4x + 6y = 14
\end{aligned}
$$

Add them to eliminate y:

$$
9x - 6y + 4x + 6y = 15 + 14 \implies 13x = 29 \implies x = \frac{29}{13}
$$

**Step 2. Substitute** $$x = \frac{29}{13}$$ back into one of the original equations (say $$3x - 2y = 5$$):

$$
3\left(\frac{29}{13}\right) - 2y = 5
$$

$$
\frac{87}{13} - 2y = 5 \implies -2y = 5 - \frac{87}{13}
$$

$$
5 = \frac{65}{13}, \text{ so } 5 - \frac{87}{13} = \frac{65}{13} - \frac{87}{13} = -\frac{22}{13}
$$

$$
-2y = -\frac{22}{13} \implies y = \frac{11}{13}
$$

Hence,

$$
(x,y) = \left(\frac{29}{13}, \frac{11}{13}\right)

<h4><b>Solution to System 4 </b></h4>

$$
\begin{cases}
2x - 3y + 4z + 2t = 2 \\
3x + 2y - 5z + 3t = 3 \\
4x - 3y + 2z - 5t = 4 \\
5x + 4y - 3z + 2t = 5
\end{cases}
$$

Label them:

$$
\begin{aligned}
(\mathrm{I}) &: 2x - 3y + 4z + 2t = 2 \\
(\mathrm{II}) &: 3x + 2y - 5z + 3t = 3 \\
(\mathrm{III}) &: 4x - 3y + 2z - 5t = 4 \\
(\mathrm{IV}) &: 5x + 4y - 3z + 2t = 5
\end{aligned}
$$

1. Combine (I) and (II) to eliminate x.

Multiply (I) by 3:

$$
6x - 9y + 12z + 6t = 6
$$

Multiply (II) by 2:

$$
6x + 4y - 10z + 6t = 6
$$

Subtract one from the other:

$$
(6x - 9y + 12z + 6t) - (6x + 4y - 10z + 6t) = 6 - 6
$$

Left side:

$$
6x - 6x = 0, \quad -9y - 4y = -13y, \quad 12z - (-10z) = 12z + 10z = 22z, \quad 6t - 6t = 0
$$

Right side: 0.

Hence

$$
-13y + 22z = 0 \implies 13y = 22z \implies y = \frac{22}{13}z \tag{A}
$$

2. Combine (I) and (III) to eliminate x.

Multiply (I) by 4:

$$
8x - 12y + 16z + 8t = 8
$$

Multiply (III) by 2:

$$
8x - 6y + 4z - 10t = 8
$$

Subtract one from the other:

$$
(8x - 12y + 16z + 8t) - (8x - 6y + 4z - 10t) = 8 - 8
$$

Left side:

$$
8x - 8x = 0, \quad -12y - (-6y) = -12y + 6y = -6y, \quad 16z - 4z = 12z, \quad 8t - (-10t) = 8t + 10t = 18t
$$

Right side: 0.

Hence

$$
-6y + 12z + 18t = 0
$$

Divide by 6:

$$
-y + 2z + 3t = 0 \implies y = 2z + 3t \tag{B}
$$

3. Combine (A) and (B) to relate z and t.
From (A):

$$
y = \frac{22}{13}z
$$

From (B):

$$
y = 2z + 3t
$$

Set them equal:

$$
2z + 3t = \frac{22}{13}z \implies 3t = \frac{22}{13}z - 2z
$$

$$
2z = \frac{26}{13}z, \quad \frac{22}{13}z - \frac{26}{13}z = -\frac{4}{13}z
$$

so

$$
3t = -\frac{4}{13}z \implies t = -\frac{4}{39}z
$$

Hence from (A) or (B):

$$
y = 2z + 3\left(-\frac{4}{39}z\right) = 2z - \frac{12}{39}z = 2z - \frac{4}{13}z

Use (I) again:

$$
2x - 3y + 4z + 2t = 2
$$

Substitute $$y = \frac{22}{13}z$$ and $$t = -\frac{4}{39}z$$:

$$
2x - 3\left(\frac{22}{13}z\right) + 4z + 2\left(-\frac{4}{39}z\right) = 2

Compute step by step:

$$
-3 \cdot \frac{22}{13}z = -\frac{66}{13}z, \quad 4z = \frac{52}{13}z

Thus

$$
-\frac{66}{13}z + \frac{52}{13}z = -\frac{14}{13}z

$$
2\left(-\frac{4}{39}z\right) = -\frac{8}{39}z
$$

Convert $$-\frac{14}{13}z$$ to denominator 39:

$$
-\frac{14}{13}z = -\frac{14 \times 3}{13 \times 3}z = -\frac{42}{39}z

So

$$
-\frac{42}{39}z - \frac{8}{39}z = -\frac{50}{39}z
$$

Thus the equation becomes:

$$
2x - \frac{50}{39}z = 2 \implies 2x = 2 + \frac{50}{39}z \implies x = 1 + \frac{25}{39}z

Equation (IV) is:

$$
5x + 4y - 3z + 2t = 5

Substitute

5 + \frac{264}{39}z=5

$$

From z = 0,

$$
t = -\frac{4}{39} \cdot 0 = 0
$$

From

$$
y = \frac{22}{13} \cdot 0 = 0
$$

From

$$
x = 1 + \frac{25}{39} \cdot 0 = 1
$$

Hence,

$$
(x,y,z,t) = (1, 0, 0, 0)
$$

## Solution: 
**Problem 2:**

The system of equations:

$$\begin{cases}
x_1 + 2x_2 + 3x_3 = 41, \\
4x_1 + 5x_2 + 6x_3 = 93, \\
7x_1 + 8x_2 + 9x_3 = 145.
\end{cases}$$

**Explanation:**

Similarly, this system can be written as:

$$A \cdot X = B,$$

where:

$$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}, \quad
X = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}, \quad
B = \begin{bmatrix} 41 \\ 93 \\ 145 \end{bmatrix}.$$

The matrix $A$ is singular, meaning it does not have an inverse. This happens because the rows of $A$ are linearly dependent. In this case:

$$\text{Row 3} = \text{Row 1} + \text{Row 2}.$$

As a result, the system does not have a unique solution. Instead:
* It could have infinitely many solutions if $B$ lies in the column space of $A$.
* It could be inconsistent (no solution) if $B$ does not lie in the column space of $A$.

Since $B = [41, 93, 145]^T$ is consistent with the linear dependence, this system likely has infinitely many solutions. To find these solutions, methods like row reduction (RREF) can be used to express the variables in terms of free parameters.
