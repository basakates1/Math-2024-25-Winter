## Solution 3:
Find the slope of the given line:

The equation of the given line is $y = 2x + 3$. The slope ($m$) of this line is $2$.

Find the slope of the perpendicular line:

The product of the slopes of two perpendicular lines is $-1$. Let $m_2$ be the slope of the perpendicular line. Then:

$$m \cdot m_2 = -1 \implies 2 \cdot m_2 = -1 \implies m_2 = -\frac{1}{2}.$$

Write the equation of the perpendicular line:

The general form of a line is:

$$y - y_1 = m_2(x - x_1),$$

where $(x_1, y_1) = (1, 2)$ and $m_2 = -\frac{1}{2}$. Substituting these values:

$$y - 2 = -\frac{1}{2}(x - 1).$$

Simplify the equation:

Distribute and rearrange:

$$y - 2 = -\frac{1}{2}x + \frac{1}{2}.$$

$$y = -\frac{1}{2}x + \frac{5}{2}.$$

Thus, the equation of the perpendicular line is:

$$y = -\frac{1}{2}x + \frac{5}{2}.$$

#### **Step 1: Find the Intersection Point**
The equations of the lines are:
$$
y = 2x + 3 \quad \text{(Line 1)}
$$
$$
y = 3x + 2 \quad \text{(Line 2)}
$$

At the intersection point, the \( y \)-values of the two lines are equal. Setting the equations equal to each other:
$$
2x + 3 = 3x + 2
$$

Simplify:
$$
3 - 2 = 3x - 2x
$$
$$
x = 1
$$

Substitute \( x = 1 \) into either line equation to find \( y \). Using \( y = 2x + 3 \):
$$
y = 2(1) + 3 = 5
$$

Thus, the intersection point is:
$$
(1, 5)
$$

---

#### **Step 2: Calculate the Angle Between the Lines**
The slopes (\( m \)) of the lines are:
- Slope of Line 1 (\( m_1 \)): \( 2 \)
- Slope of Line 2 (\( m_2 \)): \( 3 \)

The formula for the angle (\( \theta \)) between two lines is:
$$
\tan(\theta) = \left| \frac{m_2 - m_1}{1 + m_1 m_2} \right|
$$

Substitute the values:
$$
\tan(\theta) = \left| \frac{3 - 2}{1 + (2)(3)} \right| = \left| \frac{1}{1 + 6} \right| = \frac{1}{7}
$$

Find \( \theta \) by taking the arctangent:
$$
\theta = \arctan\left(\frac{1}{7}\right)
$$

Convert the result to degrees:
$$
\theta \approx \arctan(0.142857) \approx 8.13^\circ
$$

---

### **Final Answer**
1. The intersection point of the lines is:
   $$
   (1, 5)
   $$

2. The angle between the two lines is:
   $$
   \theta \approx 8.13^\circ

## Solution 5

### Problem
Find the equation of the line passing through the point \( A(1, 2) \) and parallel to the vector \( \vec{v} = [2, 3] \).

---

### Solution

#### 1. Vector Form of the Line
The vector form of a line is:
$$
\vec{r}(t) = \vec{r}_0 + t\vec{v},
$$
where:
- \( \vec{r}_0 = (1, 2) \) (point \( A \)),
- \( \vec{v} = (2, 3) \) (direction vector).

Thus, the equation in vector form is:
$$
\vec{r}(t) = (1, 2) + t(2, 3), \quad t \in \mathbb{R}.
$$

---

#### 2. Parametric Equations
The parametric equations of the line are:
$$
x = 1 + 2t, \quad y = 2 + 3t, \quad t \in \mathbb{R}.
$$

---

#### 3. Slope-Intercept Form
To convert to slope-intercept form, we eliminate \( t \) from the parametric equations.

From \( x = 1 + 2t \), we find:
$$
t = \frac{x - 1}{2}.
$$

Substitute \( t \) into \( y = 2 + 3t \):
$$
y = 2 + 3\left(\frac{x - 1}{2}\right),
$$
simplify:
$$
y = 2 + \frac{3}{2}(x - 1),
$$
$$
y = \frac{3}{2}x + \frac{1}{2}.
$$

Thus, the slope-intercept form of the equation is:
$$
y = \frac{3}{2}x + \frac{1}{2}.
$$

---

### Final Answer
The equation of the line is:
$$
y = \frac{3}{2}x + \frac{1}{2}.
$$

[![Visualization by Python](https://i.postimg.cc/j5N2syqS/output-8.png)](https://postimg.cc/LY92DYZc)


```python
#Visualization code by Python - Solution 5

import matplotlib.pyplot as plt
import numpy as np

# Define the point and vector
A = np.array([1, 2])
v = np.array([2, 3])

# Define the line (parametric form)
t = np.linspace(-5, 5, 100)  # Parameter t
x = A[0] + v[0] * t
y = A[1] + v[1] * t

# Plot the line
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Line: y = 3/2x + 1/2", color="blue")

# Plot the point A
plt.scatter(A[0], A[1], color="red", label="Point A(1, 2)")

# Plot the direction vector
plt.quiver(A[0], A[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color="green", label="Direction Vector [2, 3]")

# Enhance the plot
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.title("Line Passing Through A(1, 2) and Parallel to [2, 3]")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5, 5)
plt.ylim(-5, 10)
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.show()

```

## Solution 6

#### Given Line
The equation of the given line is:
$$ y = 2x + 3 $$

#### Parallel Line
A line parallel to the given one will have the same slope. The general equation of a parallel line is:
$$ y = 2x + c $$

For example, if \( c = -5 \), the equation of the parallel line becomes:
$$ y = 2x - 5 $$

#### Perpendicular Line
A line perpendicular to the given one will have a slope that is the negative reciprocal of the original slope. Since the slope of the given line is \( m = 2 \), the slope of the perpendicular line will be:
$$ m = -\frac{1}{2} $$

The general equation of the perpendicular line is:
$$ y = -\frac{1}{2}x + c $$

For example, if \( c = 4 \), the equation of the perpendicular line becomes:
$$ y = -\frac{1}{2}x + 4 $$

---

### Correspondence to Visualization

   $$ y = -\frac{1}{2}x + 4 $$

# Solution 7
To calculate the distance from the point \( A(1, 2) \) to the line \( y = 2x + 3 \), we use the formula for the distance from a point to a line:

$$
\text{Distance} = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

The given line \( y = 2x + 3 \) can be rewritten in standard form:

$$
2x - y + 3 = 0
$$

Here:
- \( A = 2 \)
- \( B = -1 \)
- \( C = 3 \)

The coordinates of the point are \( (x_1, y_1) = (1, 2) \).

### Step 1: Substitute into the formula
Substitute the values into the formula:

$$
\text{Distance} = \frac{|2(1) - 1(2) + 3|}{\sqrt{2^2 + (-1)^2}}
$$

### Step 2: Simplify the numerator
Calculate the numerator:

$$
|2(1) - 1(2) + 3| = |2 - 2 + 3| = |3| = 3
$$

### Step 3: Simplify the denominator
Calculate the denominator:

$$
\sqrt{2^2 + (-1)^2} = \sqrt{4 + 1} = \sqrt{5}
$$

### Step 4: Final calculation
Combine the results:

$$
\text{Distance} = \frac{3}{\sqrt{5}}
$$

To rationalize the denominator:

$$
\text{Distance} = \frac{3\sqrt{5}}{5}
$$

### Final Answer:
The exact distance is:

$$
\frac{3\sqrt{5}}{5}
$$

The approximate distance is:

$$
1.34
$$

```python
# Visalization by Python - Solution 7

import numpy as np
import matplotlib.pyplot as plt

# Define the line y = 2x + 3
x_vals = np.linspace(-2, 3, 100)
y_vals = 2 * x_vals + 3

# Define the point A(1, 2)
point_A = (1, 2)

# Calculate the distance from point A to the line
# Using the formula |Ax1 + By1 + C| / sqrt(A^2 + B^2)
A = 2  # Coefficient of x in the line equation
B = -1 # Coefficient of y in the line equation
C = 3  # Constant term in the line equation
x1, y1 = point_A

distance = abs(A * x1 + B * y1 + C) / np.sqrt(A**2 + B**2)

# Plot the line and the point
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="Line: $y = 2x + 3$", color="blue")
plt.scatter(*point_A, color="red", label="Point A(1, 2)")
plt.axhline(0, color='black',linewidth=0.5, linestyle='--')
plt.axvline(0, color='black',linewidth=0.5, linestyle='--')

# Show the distance visually by drawing a perpendicular line
x_perpendicular = (x1 + (y1 - 2*x1 - 3) / (2 + 1 / 2)) / (1 + 4)
y_perpendicular = 2 * x_perpendicular + 3
plt.plot([x1, x_perpendicular], [y1, y_perpendicular], color="green", linestyle="--", label=f"Distance = {distance:.2f}")

# Add legend and labels
plt.title("Distance from a Point to a Line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()

distance
```

## Solution 8
### Solution

The equation of the line can be determined as follows:

1. **Find the slope \( m \):**

   The slope \( m \) is calculated using the formula:
   $$
   m = \frac{y_2 - y_1}{x_2 - x_1}
   $$
   For the points \( A(2, 0) \) and \( B(0, 3) \):
   $$
   m = \frac{3 - 0}{0 - 2} = \frac{3}{-2} = -\frac{3}{2}
   $$

2. **Write the slope-intercept form:**

   The general equation of a line in slope-intercept form is:
   $$
   y = mx + c
   $$
   Substitute \( m = -\frac{3}{2} \), and use one of the points (e.g., \( A(2, 0) \)) to find \( c \).

3. **Find \( c \):**

   Substitute \( x = 2 \), \( y = 0 \), and \( m = -\frac{3}{2} \) into the equation:
   $$
   0 = -\frac{3}{2}(2) + c
   $$
   Simplify:
   $$
   0 = -3 + c \implies c = 3
   $$

4. **Write the final equation:**

   Substitute \( m = -\frac{3}{2} \) and \( c = 3 \) into the slope-intercept form:
   $$
   y = -\frac{3}{2}x + 3
   $$

### Final Answer:
The equation of the line is:
$$
y = -\frac{3}{2}x + 3
$$

```python
#Python code for visualization - Solution 8

import matplotlib.pyplot as plt
import numpy as np

# Define the points of intersection
A = (2, 0)
B = (0, 3)

# Calculate the slope (m) of the line
m = (B[1] - A[1]) / (B[0] - A[0])

# Calculate the y-intercept (c)
c = A[1] - m * A[0]

# Define the equation of the line
x = np.linspace(-1, 3, 100)
y = m * x + c

# Plotting the line and points
plt.figure(figsize=(8, 6))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.plot(x, y, label=f"Line: y = {m:.2f}x + {c:.2f}")
plt.scatter(*A, color='red', label='A(2, 0)')
plt.scatter(*B, color='blue', label='B(0, 3)')
plt.title("Graph of the Line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()

m, c
```

# Solution 9

To solve for the angle between the line \( y = x + 3 \) and the \( Ox \)-axis:

1. **Slope of the Line**:
   The equation of the line is \( y = x + 3 \), which is in the slope-intercept form \( y = mx + b \), where \( m \) (the slope) is:
   $$
   m = 1
   $$

2. **Relationship Between Slope and Angle**:
   The slope \( m \) is related to the angle \( \theta \) between the line and the \( Ox \)-axis by:
   $$
   \tan(\theta) = m
   $$

3. **Calculate the Angle**:
   Substituting \( m = 1 \):
   $$
   \tan(\theta) = 1
   $$
   The angle \( \theta \) is given by:
   $$
   \theta = \arctan(1)
   $$
   Using the property of arctangent:
   $$
   \theta = 45^\circ
   $$

Thus, the angle between the line \( y = x + 3 \) and the \( Ox \)-axis is:
$$
\boxed{45^\circ}
$$


[Solution by Python](https://postimg.cc/jCfZcgRr)

```python
#Solution 9 Python Visualization

import numpy as np
import matplotlib.pyplot as plt

# Calculate the angle between the line and the Ox axis
slope = 1  # The slope of the line y = x + 3
angle_rad = np.arctan(slope)  # Angle in radians
angle_deg = np.degrees(angle_rad)  # Convert to degrees

# Visualization
x = np.linspace(-10, 10, 100)
y = x + 3

plt.figure(figsize=(8, 6))
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.plot(x, y, label=r"$y = x + 3$", color="blue")

# Highlight the angle between the line and Ox axis
plt.quiver(0, 0, 1, 0, angles="xy", scale_units="xy", scale=1, color="red", label="Ox axis")
plt.quiver(0, 0, 1, slope, angles="xy", scale_units="xy", scale=1, color="green", label="Line direction")
plt.text(0.5, 0.2, r"$\theta = {:.2f}^\circ$".format(angle_deg), fontsize=12, color="purple")

# Labels and legend
plt.xlim(-5, 5)
plt.ylim(-5, 10)
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()
plt.title("Angle Between Line $y = x + 3$ and Ox Axis")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

angle_deg
```

## Solution  10

To find a vector perpendicular to the line \(x + y + 1 = 0\), follow these steps:

### Step 1: Rewrite the equation of the line
The given line equation is:
$$
x + y + 1 = 0
$$
Rearranging it into slope-intercept form:
$$
y = -x - 1
$$
The slope of the line is \(-1\).

### Step 2: Direction vector of the line
The direction vector of a line is derived from its slope. For a line with slope \(m = -1\), the direction vector can be written as:
$$
\vec{d} = [1, -1]
$$

### Step 3: Find a perpendicular vector
A vector \(\vec{v}\) is perpendicular to \(\vec{d}\) if their dot product is zero:
$$
\vec{d} \cdot \vec{v} = 0
$$
Substitute \(\vec{d} = [1, -1]\):
$$
[1, -1] \cdot [v_1, v_2] = 1 \cdot v_1 + (-1) \cdot v_2 = 0
$$
This simplifies to:
$$
v_1 - v_2 = 0 \quad \implies \quad v_1 = v_2
$$
Thus, a perpendicular vector can be chosen as:
$$
\vec{v} = [1, 1]
$$

### Step 4: Conclusion
A vector perpendicular to the line \(x + y + 1 = 0\) is:
$$
\boxed{\vec{v} = [1, 1]}
$$
Alternatively, \([-1, -1]\) is also valid.

```python
#Python Code for Solution 10

import matplotlib.pyplot as plt
import numpy as np

# Line equation: x + y + 1 = 0
# Rearrange to y = -x - 1
x = np.linspace(-10, 10, 400)
y = -x - 1

# Perpendicular vector: The normal vector to the line is (1, 1)
# Draw the normal vector from a point on the line (e.g., origin (-1, 0))
origin = np.array([-1, 0])
normal_vector = np.array([1, 1])

# Plotting the line and vector
plt.figure(figsize=(8, 8))
plt.plot(x, y, label="Line: x + y + 1 = 0", color="blue")
plt.quiver(*origin, *normal_vector, angles='xy', scale_units='xy', scale=1, color="red", label="Perpendicular Vector (1, 1)")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title("Visualization of the Line and Perpendicular Vector")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
