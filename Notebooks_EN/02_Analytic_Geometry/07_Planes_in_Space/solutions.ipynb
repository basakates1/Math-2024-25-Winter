<h4><b>Solution 1:</b></h4>

To find the equation of a plane passing through three points $A(1,2,3)$, $B(3,4,5)$, and $C(2,1,4)$, we'll use the following steps:

1. Find vectors in the plane: We can start by creating two vectors that lie within the plane, namely $\vec{AB}$ and $\vec{AC}$, using the points given:

$\vec{AB} = B - A = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2)$

$\vec{AC} = C - A = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1)$

2. Find a normal vector to the plane: A normal vector to the plane can be obtained by taking the cross product of $\vec{AB}$ and $\vec{AC}$. The cross product of two vectors $\vec{u} = (u_1, u_2, u_3)$ and $\vec{v} = (v_1, v_2, v_3)$ is calculated as:

$\vec{u} \times \vec{v} = (u_2 v_3 - u_3 v_2, u_3 v_1 - u_1 v_3, u_1 v_2 - u_2 v_1)$

Substituting the components of $\vec{AB}$ and $\vec{AC}$:

$\vec{AB} \times \vec{AC} = (2 \times 1 - 2 \times -1, 2 \times 1 - 2 \times 1, 2 \times -1 - 2 \times 1) = (2+2, 2-2, -2-2) = (4, 0, -4)$

3. Simplify the normal vector: The vector $(4, 0, -4)$ is a valid normal, but it can be simplified to $(1, 0, -1)$ for the purpose of forming the plane equation.

4. Formulate the plane equation: The equation of a plane given a normal vector $(A, B, C)$ and a point $(x_1, y_1, z_1)$ on the plane is:

$A(x - x_1) + B(y - y_1) + C(z - z_1) = 0$

Using the simplified normal vector $(1, 0, -1)$ and point $A(1, 2, 3)$:

$1(x - 1) + 0(y - 2) + (-1)(z - 3) = 0$

$x - 1 - z + 3 = 0$

$x - z + 2 = 0$

$x - z = -2$
or
$z = x + 2$

Hence, the equation of the plane is $x - z + 2 = 0$ or $z = x+2$.

To find the equation of the plane passing through a point and parallel to two given vectors, we can follow these steps:

1.  Find the normal vector to the plane by taking the cross product of the two parallel vectors.

2.  Use the point-normal form of the plane's equation: $(x - x_0, y - y_0, z - z_0) \cdot \vec{n} = 0$, where $\vec{n}$ is the normal vector and $(x_0, y_0, z_0)$ is a point through which the plane passes.

Let's start by finding the normal vector to the plane using the cross product of vectors $\vec{v_1} = [1, 0, 1]$ and $\vec{v_2} = [0, 1, -1]$:

$\vec{v_1} \times \vec{v_2} = \begin{vmatrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ 1 & 0 & 1 \\ 0 & 1 & -1 \end{vmatrix}$

The determinant for each component is calculated as:

*   i-component: $0 \cdot (-1) - 1 \cdot 1 = -1$
*   j-component: $1 \cdot (-1) - 1 \cdot 0 = -1$. However, remember that the j-component in the cross product has a negative sign in front of it, so it becomes $-(-1) = 1$
*   k-component: $1 \cdot 1 - 0 \cdot 0 = 1$

So, the cross product $\vec{n} = [-1, 1, 1]$.

Now, using the point-normal form of the plane's equation with point $A(1, 2, 3)$ and normal vector $\vec{n} = [-1, 1, 1]$, we have:

$-1(x - 1) + 1(y - 2) + 1(z - 3) = 0$

$-x + 1 + y - 2 + z - 3 = 0$

$-x + y + z - 4 = 0$

Hence, the equation of the plane is:

$-x + y + z = 4$ or $x - y - z = -4$

This equation represents the plane passing through the point $(1, 2, 3)$ and parallel to the vectors $\vec{v_1}$ and $\vec{v_2}$.

<h4><b>Solution 6</b></h4>

**Parallel Plane**

A plane that is parallel to another has the same normal vector. The given plane is:

$2x + 3y + 4z = 5$

The normal vector of this plane is $\mathbf{n} = (2, 3, 4)$.

Any plane with the same normal vector is parallel. We can change the constant on the right side of the equation to get a different plane:

$2x + 3y + 4z = k$

where $k \ne 5$ (to ensure it's not the same plane). Let's pick $k = 10$ as an example:

$2x + 3y + 4z = 10$

**Perpendicular Plane**

A plane that is perpendicular to another will have a normal vector that is perpendicular to the normal vector of the given plane. The dot product of the normal vectors of two perpendicular planes equals zero:

$\mathbf{n}_1 \cdot \mathbf{n}_2 = 0$

Letâ€™s assume a normal vector for the perpendicular plane as $\mathbf{n}_2 = (a, b, c)$. It must satisfy:

$2a + 3b + 4c = 0$

Choosing simple values that satisfy this, let $a = 4$, $b = -4$, and $c = 1$ (since $2(4) + 3(-4) + 4(1) = 8 - 12 + 4 = 0$), we get the normal vector $(4, -4, 1)$ for the perpendicular plane. The equation of the perpendicular plane can be written as:

$4x - 4y + z = d$

where $d$ can be any constant. Let's choose $d = 0$ for simplicity:

$4x - 4y + z = 0$

**Summary**

Parallel Plane: $2x + 3y + 4z = 10$

Perpendicular Plane: $4x - 4y + z = 0$

$d = \frac{|Ax_1 + By_1 + Cz_1 + D|}{\sqrt{A^2 + B^2 + C^2}}$

In this case, $A = 2$, $B = 3$, $C = 4$, and $D = -5$ (rewriting the plane equation in standard form $2x + 3y + 4z - 5 = 0$).

$d = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}}$

$d = \frac{|15|}{\sqrt{29}}$

$d = \frac{15}{\sqrt{29}} \approx \frac{15}{5.385} \approx 2.785$

So, the distance from point $A$ to the plane is approximately 2.785 units.

<h4><b>Solution 9</b></h4>

To find the angle between two planes, you can use the normals of the planes. The angle between the planes is the same as the angle between their normal vectors.

For the planes:

$x + y + z = 1$ has the normal vector $\mathbf{n}_1 = (1, 1, 1)$.

$x = 0$ (which is the $yz$-plane) has the normal vector $\mathbf{n}_2 = (1, 0, 0)$.

The angle $\theta$ between two vectors $\mathbf{a}$ and $\mathbf{b}$ can be found using the dot product formula:

$\cos \theta = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\| \|\mathbf{b}\|}$

Here, the dot product $\mathbf{n}_1 \cdot \mathbf{n}_2 = 1 \cdot 1 + 1 \cdot 0 + 1 \cdot 0 = 1$ and the magnitudes $\|\mathbf{n}_1\| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}$ and $\|\mathbf{n}_2\| = \sqrt{1^2} = 1$.

Thus,

$\cos \theta = \frac{1}{\sqrt{3} \cdot 1} = \frac{1}{\sqrt{3}} = \frac{\sqrt{3}}{3}$

$\theta = \cos^{-1}\left(\frac{\sqrt{3}}{3}\right)$

The angle between the plane $x + y + z = 1$ and the $yz$-plane ($x = 0$) is approximately $54.74^\circ$.
