Given:

Center $A(h,k)=(1,2)$

Radius $r=3$

The standard equation of a circle with center $(h,k)$ and radius $r$ is

$(x-h)^2+(y-k)^2=r^2.$

Substitute $h=1$, $k=2$, and $r=3$:

$(x-1)^2+(y-2)^2=3^2 \Longrightarrow (x-1)^2+(y-2)^2=9.$

Thus, the equation of the circle is

$(x-1)^2+(y-2)^2=9.$

**SOLUTION 2**

Step-by-step Solution

We want a parabola that:

*   Intersects the $x$-axis at $x=2$ and $x=4$.
*   Passes through the point $(3,1)$.

1. General Form Using Known Roots

If a parabola intersects the $x$-axis at $x=r_1$ and $x=r_2$, it means $(x-r_1)$ and $(x-r_2)$ are factors. Hence we can express it in the form:

$y=a\,(x-2)\,(x-4).$

Here $a$ is a constant that we still need to determine.

2. Imposing the Condition $y(3)=1$

We are given that when $x=3$, then $y(3)=1$. Substitute $x=3$ and $y=1$ into the general form:

$1=a\,(3-2)\,(3-4).$

Compute each factor:

$3-2=1.$
$3-4=-1.$

Thus,

$1=a\cdot 1\cdot (-1)=-a.$

Solving for $a$ gives:

$-a=1 \Longrightarrow a=-1.$

3. Final Equation of the Parabola

Plugging $a=-1$ back into $y=a\,(x-2)\,(x-4)$:

$y=-1\cdot(x-2)(x-4).$

We can also expand if desired:

$(x-2)(x-4) = x^2 - 4x - 2x + 8 = x^2 - 6x + 8.$

Hence,

$y = -(x^2 - 6x + 8) = -x^2 + 6x - 8.$

Therefore, the equation of the parabola is:

$y=-x^2+6x-8.$

**SOLUTION 3**


To find the center of the ellipse given by

$x^2 + 4y^2 - 4x - 16y + 16 = 0,$

we complete the square for $x$ and $y$.

1. Group $x$ and $y$ terms:

$x^2 - 4x + 4y^2 - 16y + 16 = 0.$

2. Complete the square for $x$:

$x^2 - 4x = x^2 - 4x + 4 - 4 = (x - 2)^2 - 4.$

3. Complete the square for $y$:

Factor out the 4 from the $y$-terms:

$4y^2 - 16y = 4(y^2 - 4y) = 4[y^2 - 4y + 4 - 4] = 4((y - 2)^2 - 4) = 4(y - 2)^2 - 16.$

4. Substitute back into the equation:

$x^2 - 4x + 4y^2 - 16y + 16 = [(x - 2)^2 - 4] + [4(y - 2)^2 - 16] + 16 = 0.$

Combine like terms:

$(x - 2)^2 - 4 + 4(y - 2)^2 - 16 + 16 = 0 \Longrightarrow (x - 2)^2 + 4(y - 2)^2 - 4 = 0.$

Thus,

$(x - 2)^2 + 4(y - 2)^2 = 4.$

5. Read off the center

An ellipse of the form $(x - h)^2 + 4(y - k)^2 = 4$ is centered at $(h, k)$. Hence, the center is

$(2, 2).$

```python
# Step-by-step computation of the ellipse center --- SOLUTION 3

# Completing the square for x terms: x^2 - 4x
x_center = -(-4) / (2 * 1)  # Center of x is -b/2a
x_constant = (x_center**2)  # Adding this term for completing the square

# Completing the square for y terms: 4y^2 - 16y
y_center = -(-16) / (2 * 4)  # Center of y is -b/2a
y_constant = (y_center**2)  # Adding this term for completing the square
y_adjustment = 4 * y_constant  # Accounting for the 4 multiplier in 4y^2

# Adjusting the equation's constant term
constant_adjustment = x_constant + y_adjustment

# The original constant was 16; adjust it
final_constant = 16 - constant_adjustment

# Center of the ellipse
ellipse_center = (x_center, y_center)

# Visualization of the ellipse
x = np.linspace(-2, 6, 400)
y = np.linspace(-2, 6, 400)
X, Y = np.meshgrid(x, y)

# Equation of the ellipse after completing the square
ellipse_equation = (X - x_center)**2 + 4 * (Y - y_center)**2 - final_constant

# Plotting
plt.contour(X, Y, ellipse_equation, levels=[0], colors='blue')
plt.scatter(x_center, y_center, color='red', label='Center')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("Ellipse and its Center")
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()

ellipse_center
```

**SOLUTION 4**

We want the line

$y=mx-5$

to be tangent to the circle

$x^2 + y^2 = 1.$

Write the line in standard form.

The given line is:

$y = mx - 5 \Longrightarrow mx - y - 5 = 0.$

Use the distance-from-center-to-line formula.

A circle centered at $(0,0)$ with radius $r=1$ will be tangent to a line if the distance from the center (which is $(0,0)$) to the line equals the radius. For the line $Ax+By+C=0$, the distance from $(0,0)$ to that line is

$\frac{|C|}{\sqrt{A^2 + B^2}}.$

In our case,

$A = m, \quad B = -1, \quad C = -5.$

So the distance from $(0,0)$ to the line $mx - y - 5 = 0$ is

$\frac{|{-5}|}{\sqrt{m^2 + (-1)^2}} = \frac{5}{\sqrt{m^2 + 1}}.$

Set the distance equal to the circle’s radius.

Since the circle’s radius is 1, we require:

$\frac{5}{\sqrt{m^2 + 1}} = 1.$

Solving for $m$:

$5 = \sqrt{m^2 + 1} \Longrightarrow m^2 + 1 = 25 \Longrightarrow m^2 = 24 \Longrightarrow m = \pm \sqrt{24}.$

Select the positive slope.

Since the problem states $m>0$,

$m = \sqrt{24} = 2\sqrt{6}.$

Final Answer

$2\sqrt{6}$

The hyperbola given is

$x^2 - y^2 = 1.$

This is a standard form of a hyperbola centered at the origin with its transverse axis along the $x$-axis. It has two branches:

*   Right branch: $x > 0$,
*   Left branch: $x < 0$.

Typically, when asked for the distance between the branches of such a hyperbola, one is referring to the shortest distance between any point on the left branch and any point on the right branch.

1. Parametric or implicit approach

A point $(x,y)$ on the hyperbola satisfies

$x^2 - y^2 = 1 \Longrightarrow x = \pm\sqrt{1 + y^2}.$

Points on the right branch can be written as $(+\sqrt{1 + y^2},\, y)$.
Points on the left branch can be written as $(-\sqrt{1 + y^2},\, y)$.

Let

$P = \bigl(-\sqrt{1 + y^2},\, y\bigr) \quad\text{(on the left branch)},$
$Q = \bigl(+\sqrt{1 + y^2},\, y\bigr) \quad\text{(on the right branch)}.$

The distance $d$ between these two points (for the same $y$) is

$d(y) = \sqrt{\Bigl(\sqrt{1 + y^2} - \bigl(-\sqrt{1 + y^2}\bigr)\Bigr)^2 + (y - y)^2}.$

Since $y - y = 0$,

$d(y) = \sqrt{\Bigl(\sqrt{1 + y^2} + \sqrt{1 + y^2}\Bigr)^2} = 2\,\sqrt{1 + y^2}.$

To find the shortest such distance, minimize $d(y) = 2\sqrt{1 + y^2}$ with respect to $y$. Observe that $\sqrt{1 + y^2}$ is smallest when $y = 0$. Therefore,

$\min_{y} \, d(y) = d(0) = 2\sqrt{1 + 0^2} = 2.$

2. Geometric interpretation

The hyperbola has vertices at $(+1,\,0)$ and $(-1,\,0)$.
These two vertices lie on opposite branches of the hyperbola.
The line segment connecting $(-1,0)$ and $(+1,0)$ is simply of length 2.
Moreover, it can be checked that there is no chord connecting the left branch to the right branch that is shorter than the chord through the vertices. Hence, the distance between the two branches is $2.$
