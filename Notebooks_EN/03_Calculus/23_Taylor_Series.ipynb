# Solution 1

**1. Taylor Series for $f(x) = \cos(x)$ around $x = 0$ up to the 4th degree**

The Taylor series for a function $f(x)$ around $x = a$ is given by:

$f(x) = f(a) + f'(a)(x-a) + \frac{f''(a)}{2!}(x-a)^2 + \frac{f'''(a)}{3!}(x-a)^3 + \frac{f''''(a)}{4!}(x-a)^4 + \cdots$

For $f(x) = \cos(x)$ around $x = 0$:

* $f(x) = \cos(x)$, so $f(0) = 1$
* $f'(x) = -\sin(x)$, so $f'(0) = 0$
* $f''(x) = -\cos(x)$, so $f''(0) = -1$
* $f'''(x) = \sin(x)$, so $f'''(0) = 0$
* $f''''(x) = \cos(x)$, so $f''''(0) = 1$

Substituting into the Taylor series formula:

$\cos(x) \approx 1 + 0\cdot x + \frac{-1}{2!}x^2 + \frac{0}{3!}x^3 + \frac{1}{4!}x^4$

Simplifying:

$\cos(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}$


**2. Taylor Series for $h(x) = \frac{1}{1-x}$ around $x = 0$ up to the 4th degree**

For $h(x) = \frac{1}{1-x}$, the derivatives are:

* $h(x) = \frac{1}{1-x}$, so $h(0) = 1$
* $h'(x) = \frac{1}{(1-x)^2}$, so $h'(0) = 1$
* $h''(x) = \frac{2}{(1-x)^3}$, so $h''(0) = 2$
* $h'''(x) = \frac{6}{(1-x)^4}$, so $h'''(0) = 6$
* $h''''(x) = \frac{24}{(1-x)^5}$, so $h''''(0) = 24$

Substituting into the Taylor series formula:

$h(x) \approx h(0) + h'(0)x + \frac{h''(0)}{2!}x^2 + \frac{h'''(0)}{3!}x^3 + \frac{h''''(0)}{4!}x^4$

$h(x) \approx 1 + x + \frac{2}{2!}x^2 + \frac{6}{3!}x^3 + \frac{24}{4!}x^4$

Simplifying:

$h(x) \approx 1 + x + x^2 + x^3 + x^4$


**3. Taylor Series for $g(x) = \sin(x)$ around $x = \pi$ up to the 4th degree**

For $g(x) = \sin(x)$, the derivatives are:

* $g(x) = \sin(x)$, so $g(\pi) = 0$
* $g'(x) = \cos(x)$, so $g'(\pi) = -1$
* $g''(x) = -\sin(x)$, so $g''(\pi) = 0$
* $g'''(x) = -\cos(x)$, so $g'''(\pi) = 1$
* $g''''(x) = \sin(x)$, so $g''''(π) = 0$

Substituting into the Taylor series formula:

$g(x) \approx g(\pi) + g'(\pi)(x-\pi) + \frac{g''(\pi)}{2!}(x-\pi)^2 + \frac{g'''(\pi)}{3!}(x-\pi)^3 + \frac{g''''(π)}{4!}(x-\pi)^4$

$g(x) \approx 0 + (-1)(x-\pi) + \frac{0}{2!}(x-\pi)^2 + \frac{1}{3!}(x-\pi)^3 + \frac{0}{4!}(x-\pi)^4$

Simplifying:

$g(x) \approx -(x-\pi) + \frac{(x-\pi)^3}{6}$

# Solution 2

**Finding the Tangent Line**

**Step 1: Compute $f(x_0)$**

Evaluate $f(x)$ at $x_0 = \pi$:

$f(\pi) = e^{\sin(\pi)} = e^0 = 1$

**Step 2: Compute the derivative $f'(x)$**

Use the chain rule to differentiate $f(x) = e^{\sin(x)}$:

$f'(x) = \frac{d}{dx}(e^{\sin(x)}) = e^{\sin(x)} \cdot \cos(x)$

**Step 3: Evaluate $f'(x_0)$**

Evaluate $f'(x)$ at $x_0 = \pi$:

$f'(\pi) = e^{\sin(\pi)} \cdot \cos(\pi) = e^0 \cdot (-1) = -1$

**Step 4: Write the tangent line equation**

Using the point-slope form of a line, the tangent line at $x_0 = \pi$ is:

$y = f'(\pi)(x - \pi) + f(\pi)$

Substitute $f'(\pi) = -1$ and $f(\pi) = 1$:

$y = -1 \cdot (x - \pi) + 1$

Simplify:

$y = -x + \pi + 1$

**Final Answer:**

The tangent line to $f(x) = e^{\sin(x)}$ at $x_0 = \pi$ is:

$y = -x + \pi + 1$
