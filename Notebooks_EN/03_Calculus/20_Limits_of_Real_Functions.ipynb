## 20. Limits of Real Functions

1. Compute:
   - $\displaystyle\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$

2. Find:
   
   - $\displaystyle \lim_{x \to 0} \frac{\sin(3x)}{2x+1}$.

4. Find the asymptotes of the function:
  
   - $f(x) = \frac{x^2 - 1}{x^2 + 1}$
   - $g(x) = \frac{\sin(x)}{x^2+1}$

# Solution 1

To compute the limit:

$\lim_{x\to\infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$

**Step 1: Identify the highest degree terms**

In the numerator, the highest degree term is $x^3$.

In the denominator, the highest degree term is $x^4$.

**Step 2: Divide both the numerator and the denominator by the highest degree term in the denominator ($x^4$)**

$\lim_{x\to\infty} \frac{\frac{x^3}{x^4} + \frac{2x^2}{x^4}}{\frac{x^4}{x^4} - \frac{3x^3}{x^4}} = \lim_{x\to\infty} \frac{\frac{1}{x} + \frac{2}{x^2}}{1 - \frac{3}{x}}$

**Step 3: Simplify the expression**

As $x$ approaches infinity, $\frac{1}{x}$, $\frac{2}{x^2}$, and $\frac{3}{x}$ all approach 0.

$\lim_{x\to\infty} \frac{0 + 0}{1 - 0} = \frac{0}{1} = 0$

**Final Answer:**

0

# Solution 2

To evaluate the limit:

$\lim_{x\to 0} \frac{\sin(3x)}{2x + 1}$

**Direct Substitution:**

First, try substituting $x = 0$ directly into the expression:

$\frac{\sin(3\cdot 0)}{2\cdot 0 + 1} = \frac{\sin(0)}{1} = \frac{0}{1} = 0$

Since the denominator $2x + 1$ is not zero at $x = 0$ (it equals 1), the limit exists and is equal to 0.

**Verification Using Limit Properties:**

Alternatively, we can use the fact that $\sin(3x)$ approaches 0 as $x$ approaches 0, and the denominator $2x + 1$ approaches 1. Therefore, the entire fraction approaches:

$\frac{0}{1} = 0$

This confirms our earlier result.

**Using L'H么pital's Rule (if necessary):**

Although not required in this case, L'H么pital's Rule can be applied when both the numerator and denominator approach 0 or both approach infinity. Here, since the numerator approaches 0 and the denominator approaches 1, L'H么pital's Rule is *not* applicable.

However, for completeness, if we had a similar limit where both numerator and denominator approach 0, such as:

$\lim_{x\to 0} \frac{\sin(3x)}{2x}$

we *could* apply L'H么pital's Rule by differentiating the numerator and the denominator:

$\lim_{x\to 0} \frac{3\cos(3x)}{2} = \frac{3\cdot 1}{2} = \frac{3}{2}$

But this is not needed for the original problem.

**Final Answer:**

0

# Solution 3

**Problem 1:** $f(x) = \frac{x^2 - 1}{x^2 + 1}$

**Step 1: Identify Vertical Asymptotes**

Vertical asymptotes occur when the denominator is equal to zero.  We set the denominator equal to zero and solve for $x$:

$x^2 + 1 = 0$

$x^2 = -1$

This equation has no real solutions. Therefore, there are no vertical asymptotes for this function.

**Step 2: Identify Horizontal Asymptotes**

Horizontal asymptotes are determined by comparing the degrees of the numerator and denominator.

* Numerator Degree: 2 (from $x^2$)
* Denominator Degree: 2 (from $x^2$)

Since the degrees are the same, the horizontal asymptote is the ratio of the leading coefficients.  Both leading coefficients are 1.

Horizontal Asymptote $= \frac{1}{1} = 1$

So, there's a horizontal asymptote at $y = 1$.

**Step 3: Check for Oblique Asymptotes**

Oblique asymptotes occur when the degree of the numerator is exactly one greater than the degree of the denominator.  Since the degrees of the numerator and denominator are equal, there are no oblique asymptotes.

**Summary for $f(x)$:**

* Vertical Asymptotes: None
* Horizontal Asymptotes: $y = 1$
* Oblique Asymptotes: None

**Problem 2:** $g(x) = \frac{\sin(x)}{x^2 + 1}$

**Step 1: Identify Vertical Asymptotes**

We set the denominator equal to zero to find potential vertical asymptotes:

$x^2 + 1 = 0$

$x^2 = -1$

This equation has no real solutions. Therefore, there are no vertical asymptotes for $g(x)$.

**Step 2: Identify Horizontal Asymptotes**

To find horizontal asymptotes, we consider the behavior of $g(x)$ as $x$ approaches infinity ($\infty$) and negative infinity ($-\infty$).

The numerator, $\sin(x)$, oscillates between -1 and 1. The denominator, $x^2 + 1$, grows without bound as $x$ approaches $\pm\infty$.

Thus, as $x$ becomes very large (positively or negatively), the denominator dominates, and the value of $g(x)$ approaches zero.

$\lim_{x\to\pm\infty} \frac{\sin(x)}{x^2 + 1} = 0$

Therefore, there's a horizontal asymptote at $y = 0$.

**Step 3: Check for Oblique Asymptotes**

Oblique asymptotes occur when the degree of the numerator is exactly one more than the degree of the denominator. In this case:

* Numerator: The sine function does not have a polynomial degree, but its amplitude is bounded.
* Denominator Degree: 2

Since the numerator does not have a higher degree than the denominator, there are no oblique asymptotes for $g(x)$.

**Summary for $g(x)$:**

* Vertical Asymptotes: None
* Horizontal Asymptotes: $y = 0$
* Oblique Asymptotes: None

**Final Answer**

For $f(x) = \frac{x^2 - 1}{x^2 + 1}$:

* Vertical Asymptotes: None
* Horizontal Asymptotes: $y = 1$
* Oblique Asymptotes: None

For $g(x) = \frac{\sin(x)}{x^2 + 1}$:

* Vertical Asymptotes: None
* Horizontal Asymptotes: $y = 0$
* Oblique Asymptotes: None
