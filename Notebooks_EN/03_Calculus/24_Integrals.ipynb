# Solution 1

**Indefinite Integrals**

1. $\int 1 \, dx = x + C$
   * Explanation: The integral of a constant 1 with respect to $x$ is $x$, plus the constant of integration $C$.

2. $\int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C$
   * Explanation: Integrate term by term. The integral of $x^2$ is $\frac{x^3}{3}$, and the integral of 2 is $2x$.

3. $\int 2\sin(x) \, dx = -2\cos(x) + C$
   * Explanation: The integral of $\sin(x)$ is $-\cos(x)$, and the constant multiple 2 is factored out.

4. $\int \frac{3}{x} \, dx = 3\ln|x| + C$
   * Explanation: The integral of $\frac{1}{x}$ is $\ln|x|$, and the constant multiple 3 is factored out.

5. $\int \frac{1}{x^2} \, dx = -\frac{1}{x} + C$
   * Explanation: Rewrite $\frac{1}{x^2}$ as $x^{-2}$ and apply the power rule for integration: $\int x^n \, dx = \frac{x^{n+1}}{n+1}$ for $n \ne -1$.

6. $\int \left(\frac{1}{3}x^4 - 5\right) \, dx = \frac{1}{15}x^5 - 5x + C$
   * Explanation: Integrate term by term. The integral of $\frac{1}{3}x^4$ is $\frac{1}{15}x^5$, and the integral of $-5$ is $-5x$.

7. $\int (\sin^2 x + \cos^2 x) \, dx = \int 1 \, dx = x + C$
   * Explanation: Use the Pythagorean identity $\sin^2 x + \cos^2 x = 1$. The integral of 1 is $x$.

8. $\int (5\sin x + 3e^x) \, dx = -5\cos(x) + 3e^x + C$
   * Explanation: Integrate term by term. The integral of $5\sin(x)$ is $-5\cos(x)$, and the integral of $3e^x$ is $3e^x$.

9. $\int \sqrt[3]{x} \, dx = \int x^{1/3} \, dx = \frac{3}{4}x^{4/3} + C$
   * Explanation: Rewrite $\sqrt[3]{x}$ as $x^{1/3}$ and apply the power rule for integration.

10. $\int \sqrt{10x} \, dx = \int \sqrt{10}x^{1/2} \, dx = \frac{2\sqrt{10}}{3}x^{3/2} + C$
    * Explanation: Rewrite $\sqrt{10x}$ as $\sqrt{10}x^{1/2}$ and apply the power rule for integration.

11. $\int \cos\left(\frac{5}{2}x + 3\right) \, dx = \frac{2}{5}\sin\left(\frac{5}{2}x + 3\right) + C$
    * Explanation: Use the substitution $u = \frac{5}{2}x + 3$, $du = \frac{5}{2}dx$. The integral becomes $\frac{2}{5}\sin(u) + C$.

12. $\int \frac{\cos(\ln(x))}{x} \, dx = \sin(\ln(x)) + C$
    * Explanation: Use the substitution $u = \ln(x)$, $du = \frac{1}{x}dx$. The integral becomes $\int \cos(u) \, du = \sin(u) + C$.

13. $\int x\ln(x) \, dx = \frac{x^2}{2}\ln(x) - \frac{x^2}{4} + C$
    * Explanation: Use integration by parts with $u = \ln(x)$ and $dv = x \, dx$. Then $du = \frac{1}{x}dx$ and $v = \frac{x^2}{2}$. Apply the integration by parts formula: $\int u \, dv = uv - \int v \, du$.

14. $\int xe^x \, dx = xe^x - e^x + C$
    * Explanation: Use integration by parts with $u = x$ and $dv = e^x \, dx$. Then $du = dx$ and $v = e^x$. Apply the integration by parts formula: $\int u \, dv = uv - \int v \, du$.

# Solution 2

**1. Definite Integral of $f(x) = 2x + 1$**

$\int_0^\pi (2x + 1) \, dx$

Split the integral:

$\int_0^\pi 2x \, dx + \int_0^\pi 1 \, dx$

Compute each part:

$\int 2x \, dx = x^2 + C$,  $\int 1 \, dx = x + C$

Evaluate from $0$ to $\pi$:

$[x^2]_0^\pi + [x]_0^\pi = (\pi^2 - 0) + (\pi - 0) = \pi^2 + \pi$

**Final result:**

$\int_0^\pi (2x + 1) \, dx = \pi^2 + \pi$


**2. Definite Integral of $g(x) = x^2$**

$\int_0^\pi x^2 \, dx$

Compute the integral:

$\int x^2 \, dx = \frac{x^3}{3} + C$

Evaluate from $0$ to $\pi$:

$\left[\frac{x^3}{3}\right]_0^\pi = \frac{\pi^3}{3} - 0 = \frac{\pi^3}{3}$

**Final result:**

$\int_0^\pi x^2 \, dx = \frac{\pi^3}{3}$

# Solution 3

**Step 1: Identify the region**

The region is bounded:

* Vertically by $x = 1$ and $x = 2$,
* Horizontally by $y = 0$ (the $x$-axis) and $y = x^2 + 1$.

This is a vertically simple region, so we can integrate with respect to $x$.

**Step 2: Set up the integral**

The area $A$ is given by the integral of the difference between the upper and lower boundaries:

$A = \int_1^2 (x^2 + 1 - 0) \, dx$

Simplify the integrand:

$A = \int_1^2 (x^2 + 1) \, dx$

**Step 3: Evaluate the integral**

Compute the antiderivative of $x^2 + 1$:

$\int (x^2 + 1) \, dx = \frac{x^3}{3} + x + C$

Now evaluate from $x = 1$ to $x = 2$:

$A = \left(\frac{2^3}{3} + 2\right) - \left(\frac{1^3}{3} + 1\right)$

Simplify:

$A = \left(\frac{8}{3} + 2\right) - \left(\frac{1}{3} + 1\right)$

$A = \left(\frac{8}{3} + \frac{6}{3}\right) - \left(\frac{1}{3} + \frac{3}{3}\right)$

$A = \frac{14}{3} - \frac{4}{3}$

$A = \frac{10}{3}$

**Step 4: Final answer**

The area of the region is $\frac{10}{3}$.

# Solution 4

**Step 1: Find the antiderivative of $\sin(x)$**

The antiderivative of $\sin(x)$ is $-\cos(x)$, because:

$\frac{d}{dx}(-\cos(x)) = \sin(x)$

**Step 2: Evaluate the antiderivative at the bounds**

Substitute the upper and lower bounds into the antiderivative:

$P = [-\cos(x)]_0^\pi$

$P = -\cos(\pi) - (-\cos(0))$

**Step 3: Simplify**

We know that $\cos(\pi) = -1$ and $\cos(0) = 1$, so:

$P = -(-1) - (-1) = 1 + 1 = 2$

**Final Answer:**

The area under the sine curve over the interval $[0, \pi]$ is 2.

# Solution 5

**Step 1: Trapezoidal rule**

The trapezoidal rule approximates the integral of a function f(x) over an interval [a,b] by dividing the interval into n subintervals and summing the areas of trapezoids. The formula is:

$$\int_a^b f(x) \, dx \approx \frac{h}{2}[f(x_0)+2f(x_1)+2f(x_2)+\cdots+2f(x_{n-1})+f(x_n)]$$

where

$$h = \frac{b-a}{n}$$

is the width of each subinterval,

$$x_0 = a, \quad x_1 = a+h, \quad x_2 = a+2h, \quad ..., \quad x_n = b$$

**Step 2: Apply the trapezoidal rule**

Let $$f(x)=1+\cos^2(x)$$ and choose $$n=1000$$ for a reasonably accurate approximation. Then:

$$h = \frac{\pi-0}{1000} = \frac{\pi}{1000}$$

The integral becomes:

$$L \approx \frac{h}{2}[f(0)+2f(\frac{\pi}{1000})+2f(\frac{2\pi}{1000})+\cdots+2f(\frac{999\pi}{1000})+f(\pi)]$$

**Step 3: Compute f(x) at key points**

At: $$x=0$$

$$f(0)=1+\cos^2(0)=1+1=2$$

At: $$x=\pi$$

$$f(\pi)=1+\cos^2(\pi)=1+1=2$$

At other points: $$x=\frac{k\pi}{1000}$$ for $$k=1,2,\dots,999$$

$$f(\frac{k\pi}{1000})=1+\cos^2(\frac{k\pi}{1000})$$

**Step 4: Sum the terms**

Using a loop or a computational tool (e.g., Python, Excel, or a calculator), compute the sum:

$$Sum=f(0)+2f(\frac{\pi}{1000})+2f(\frac{2\pi}{1000})+\cdots+2f(\frac{999\pi}{1000})+f(\pi)$$

**Step 5: Final approximation**

Multiply the sum by: $$\frac{h}{2}$$

$$L \approx \frac{\pi}{2000} \cdot Sum$$

**Step 6: Numerical result**

Using $$n=1000$$, the numerical approximation gives:

$$L \approx 3.8202$$

Conclusion

The length of the sine curve $$y=\sin(x)$$ over the interval $$[0,\pi]$$ is approximately:

$$L \approx 3.8202$$

# Solution 6

**Finding Total Distance Traveled**

**Step 1: Find the Velocity Function**

The velocity $v(t)$ is the derivative of the position function $x(t)$:

$v(t) = \frac{dx}{dt} = \frac{d}{dt}(3t^2 - 6t + 1) = 6t - 6$

**Step 2: Determine When the Particle Changes Direction**

The particle changes direction when the velocity is zero. Set $v(t) = 0$ and solve for $t$:

$6t - 6 = 0 \implies t = 1$

So, the particle changes direction at $t = 1$.

**Step 3: Calculate the Distance Traveled in Each Interval**

We need to calculate the distance traveled in the intervals $[0, 1]$ and $[1, 2]$ separately and then sum them up.

From $t = 0$ to $t = 1$:

$x(0) = 3(0)^2 - 6(0) + 1 = 1$

$x(1) = 3(1)^2 - 6(1) + 1 = 3 - 6 + 1 = -2$

The distance traveled is:

$|x(1) - x(0)| = |-2 - 1| = 3$

From $t = 1$ to $t = 2$:

$x(1) = -2$ (from above)

$x(2) = 3(2)^2 - 6(2) + 1 = 12 - 12 + 1 = 1$

The distance traveled is:

$|x(2) - x(1)| = |1 - (-2)| = 3$

**Step 4: Sum the Distances**

Total distance traveled is the sum of the distances in each interval:

Total Distance $= 3 + 3 = 6$
