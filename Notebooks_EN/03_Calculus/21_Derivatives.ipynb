# Solution 1

**Derivatives**

* $y(x) = -3x + 3$ <br>
   $y'(x) = -3$

* $y(x) = \pi x + \sin(1)$ <br>
   $y'(x) = \pi$

* $y(x) = 4 + \sin(2)$ <br>
   $y'(x) = 0$

* $y(x) = 2x^3 - 3x^2 + 8x - 9$ <br>
   $y'(x) = 6x^2 - 6x + 8$

* $y(x) = 6x^{1/3}$ <br>
   $y'(x) = 6 \cdot \frac{1}{3} x^{-2/3} = 2x^{-2/3}$

* $y(x) = \sqrt{x} = x^{1/2}$ <br>
   $y'(x) = \frac{1}{2} x^{-1/2}$

* $y(x) = \cos(x) + \sin(x)$ <br>
   $y'(x) = -\sin(x) + \cos(x)$

* $y(x) = 2\sin(x)\cos(x) = \sin(2x)$  (Simplified first) <br>
   $y'(x) = 2\cos(2x)$

* $y(x) = x\sin(x)$ <br>
   $y'(x) = \sin(x) + x\cos(x)$  (Product rule)

* $y(x) = (x+1)(x+1) = (x+1)^2$ (Simplified first) <br>
   $y'(x) = 2(x+1)$

* $y(x) = \frac{x}{x+1}$ <br>
   $y'(x) = \frac{(1)(x+1) - x(1)}{(x+1)^2} = \frac{1}{(x+1)^2}$  (Quotient rule)

* $y(x) = (x+1)e^x$ <br>
   $y'(x) = e^x + (x+1)e^x = (x+2)e^x$  (Product rule)

* $y(x) = \sin(x^2)$ <br>
   $y'(x) = 2x\cos(x^2)$  (Chain rule)

* $y(x) = e^{-2x}$ <br>
   $y'(x) = -2e^{-2x}$  (Chain rule)

* $y(x) = \frac{1}{\sin(x+1)} = \csc(x+1)$  <br>
   $y'(x) = -\csc(x+1)\cot(x+1)$

* $y(x) = \sqrt{2x+1} = (2x+1)^{1/2}$ <br>
   $y'(x) = \frac{1}{2}(2x+1)^{-1/2} \cdot 2 = \frac{1}{\sqrt{2x+1}}$ (Chain Rule)

# Solution 2

To prove that $\frac{d}{dx}(\ln(\sin(x))) = \cot(x)$, we'll use the chain rule from calculus. Here's a step-by-step solution:

1. **Start with the function:**

$y = \ln(\sin(x))$

2. **Apply the chain rule:**

The chain rule states that if $y = \ln(u)$ and $u = \sin(x)$, then:

$\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx}$

3. **Compute $\frac{dy}{du}$:**

Since $y = \ln(u)$, the derivative of $y$ with respect to $u$ is:

$\frac{dy}{du} = \frac{1}{u}$

4. **Compute $\frac{du}{dx}$:**

Since $u = \sin(x)$, the derivative of $u$ with respect to $x$ is:

$\frac{du}{dx} = \cos(x)$

5. **Combine the results using the chain rule:**

$\frac{dy}{dx} = \frac{1}{u} \cdot \cos(x) = \frac{1}{\sin(x)} \cdot \cos(x)$

6. **Simplify the expression:**

$\frac{dy}{dx} = \frac{\cos(x)}{\sin(x)} = \cot(x)$

Therefore, we have shown that:

$\frac{d}{dx}(\ln(\sin(x))) = \cot(x)$

# Solution 3

To verify that $f''(x) = -f(x)$ for the function $f(x) = \cos(x)$, we'll follow these steps:

1. **Find the first derivative $f'(x)$:**

The derivative of $\cos(x)$ with respect to $x$ is:

$f'(x) = -\sin(x)$

2. **Find the second derivative $f''(x)$:**

Now, take the derivative of $f'(x)$:

$f''(x) = \frac{d}{dx}[-\sin(x)] = -\cos(x)$

3. **Compare $f''(x)$ with $-f(x)$:**

We have:

$f''(x) = -\cos(x)$

and

$-f(x) = -\cos(x)$

Therefore,

$f''(x) = -f(x)$

**Conclusion:**

We have verified that for $f(x) = \cos(x)$, the second derivative satisfies the equation:

$f''(x) = -f(x)$

* **Step 1: Check the form of the limit.**

As $x \to 0$, $\sin x \to 0$ and $x \to 0$, so the limit is of the form $\frac{0}{0}$, which is indeterminate. Thus, L'Hôpital's Rule applies.

* **Step 2: Differentiate the numerator and denominator.**

$\frac{d}{dx}(\sin x) = \cos x$, and $\frac{d}{dx}(x) = 1$.

* **Step 3: Apply L'Hôpital's Rule.**

$\lim_{x\to 0} \frac{\sin x}{x} = \lim_{x\to 0} \frac{\cos x}{1}$.

* **Step 4: Evaluate the limit.**

As $x \to 0$, $\cos x \to 1$, so:

$\lim_{x\to 0} \frac{\cos x}{1} = 1$.

**Final Answer:** 1

**2. $\lim_{x\to \infty} \frac{\ln x}{x}$**

* **Step 2: Differentiate the numerator and denominator.**

$\frac{d}{dx}(\ln x) = \frac{1}{x}$, and $\frac{d}{dx}(x) = 1$.

* **Step 3: Apply L'Hôpital's Rule.**

$\lim_{x\to \infty} \frac{\ln x}{x} = \lim_{x\to \infty} \frac{\frac{1}{x}}{1}$.

* **Step 4: Simplify and evaluate the limit.**

$\lim_{x\to \infty} \frac{1}{x} = 0$.

**Final Answer:** 0

**3. $\lim_{x\to \infty} \frac{e^x}{x}$**

# Solution 5

To find the velocity and acceleration of the particle at time $t=2$, we'll follow these steps:

**1. Given the position function:**

$x(t) = 3t^2 - 6t + 1$

**2. Find the velocity $V(t)$ by taking the first derivative of $x(t)$ with respect to $t$:**

$V(t) = \frac{dx}{dt} = \frac{d}{dt}(3t^2 - 6t + 1)$

$V(t) = 6t - 6$

**3. Find the acceleration $a(t)$ by taking the derivative of $V(t)$ with respect to $t$:**

$a(t) = \frac{dV}{dt} = \frac{d}{dt}(6t - 6)$

$a(t) = 6$

**4. Evaluate the velocity and acceleration at $t=2$:**

* **Velocity at $t=2$:**

$V(2) = 6(2) - 6 = 12 - 6 = 6 \, \text{m/s}$

* **Acceleration at $t=2$:**

$a(2) = 6 \, \text{m/s}^2$

**Final Answers:**

* Velocity at $t=2$: $6 \, \text{m/s}$
* Acceleration at $t=2$: $6 \, \text{m/s}^2$
