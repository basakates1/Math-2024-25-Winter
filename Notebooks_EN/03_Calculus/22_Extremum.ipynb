# Solution 6

**Step 1: Identify the Type of Function**

The profit function is a quadratic function of the form:

$P(u) = au^2 + bu + c$

where:

* $a = -2$
* $b = 50$
* $c = -300$

Since the coefficient of $u^2$ ($a = -2$) is negative, the parabola opens downwards. This means the function has a maximum point at its vertex.

**Step 2: Find the Vertex of the Parabola**

The vertex of a quadratic function $P(u) = au^2 + bu + c$ occurs at:

$u = -\frac{b}{2a}$

Plugging in the values of $a$ and $b$:

$u = -\frac{50}{2(-2)} = -\frac{50}{-4} = 12.5$

**Step 3: Interpret the Result**

Since the number of units sold $u$ must be a whole number, we round $12.5$ to the nearest whole numbers: $u = 12$ and $u = 13$.

**Step 4: Calculate Profit for $u=12$ and $u=13$**

* For $u = 12$:

$P(12) = -2(12)^2 + 50(12) - 300 = -2(144) + 600 - 300 = -288 + 600 - 300 = 12$

* For $u = 13$:

$P(13) = -2(13)^2 + 50(13) - 300 = -2(169) + 650 - 300 = -338 + 650 - 300 = 12$

Both $u=12$ and $u=13$ yield the same maximum profit of 12. Since the vertex occurs at $u = 12.5$, and both adjacent integers give the same profit, either 12 or 13 units can be considered the number that maximizes profit.

**Final Answer**

12.5 units (mathematically). However, since units must be whole numbers, selling either 12 or 13 units maximizes profit at 12.

# Solution 7

**Maximizing the Area of a Rectangle**

**1. Understand the Problem:**

The perimeter of the rectangle is given as 10 meters. We need to maximize the area of the rectangle.

**2. Formulate the Equations:**

Let the length and width of the rectangle be $L$ and $W$ respectively.

The perimeter $P$ of a rectangle is given by:

$P = 2L + 2W$

Given $P = 10$ meters:

$2L + 2W = 10$

Simplifying:

$L + W = 5$  (Equation 1)

The area $A$ of the rectangle is:

$A = L \times W$

**3. Express Area in Terms of One Variable:**

From Equation 1, we can express $W$ in terms of $L$:

$W = 5 - L$

Substitute this into the area formula:

$A = L \times (5 - L) = 5L - L^2$

**4. Find the Maximum Area:**

The equation $A = -L^2 + 5L$ is a quadratic equation that opens downward (since the coefficient of $L^2$ is negative), which means it has a maximum point.

The vertex of a parabola $y = ax^2 + bx + c$ occurs at $x = -\frac{b}{2a}$.

Here, $a = -1$ and $b = 5$, so:

$L = -\frac{5}{2(-1)} = \frac{5}{2} = 2.5$ meters

Substitute $L = 2.5$ back into Equation 1 to find $W$:

$W = 5 - 2.5 = 2.5$ meters

**Conclusion:**

The rectangle that encloses the largest possible area with a perimeter of 10 meters is a square with each side measuring 2.5 meters.

**Final Answer:**

2.5 meters $\times$ 2.5 meters

# Solution 8

**Step 1: Find the First Derivative**

The first derivative of $f(x)$ with respect to $x$ gives the slope of the tangent to the curve at any point $x$. For the function $f(x) = x^2 + 3x - 5$:

$f'(x) = \frac{d}{dx}(x^2) + \frac{d}{dx}(3x) + \frac{d}{dx}(-5) = 2x + 3$

**Step 2: Set the First Derivative to Zero**

To find critical points where the function could have a local maximum or minimum, set the first derivative equal to zero:

$2x + 3 = 0$

Solving for $x$:

$2x = -3$

$x = -\frac{3}{2}$

**Step 3: Determine the Nature of the Extremum**

To determine whether this critical point is a maximum or a minimum, we can examine the second derivative of $f(x)$.

$f''(x) = \frac{d}{dx}(2x + 3) = 2$

Since $f''(x) = 2 > 0$, the function is concave upwards at $x = -\frac{3}{2}$, indicating that this point is a local minimum.

**Step 4: Find the Minimum Value**

Substitute $x = -\frac{3}{2}$ back into the original function to find the minimum value:

$f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5$

Calculate each term:

$\left(-\frac{3}{2}\right)^2 = \frac{9}{4}$

$3 \times \left(-\frac{3}{2}\right) = -\frac{9}{2}$

Now, combine these:

$f\left(-\frac{3}{2}\right) = \frac{9}{4} - \frac{9}{2} - 5$

Convert all terms to have a common denominator of 4:

$\frac{9}{4} - \frac{18}{4} - \frac{20}{4} = \frac{9 - 18 - 20}{4} = -\frac{29}{4}$

**Final Answer**

The function $f(x) = x^2 + 3x - 5$ has a minimum value of $-\frac{29}{4}$ at $x = -\frac{3}{2}$.

# Solution 9

**1. Simplify the Function:**

$f(x) = \frac{x^2 + 2x + 1}{x - 1} = \frac{(x+1)^2}{x-1}$

**2. Find the Derivative:**

Using the quotient rule:

$f'(x) = \frac{(x-1)\frac{d}{dx}[(x+1)^2] - (x+1)^2\frac{d}{dx}(x-1)}{(x-1)^2}$

$f'(x) = \frac{(x-1)(2(x+1)) - (x+1)^2(1)}{(x-1)^2}$

$f'(x) = \frac{2(x-1)(x+1) - (x+1)^2}{(x-1)^2}$

$f'(x) = \frac{(x+1)[2(x-1) - (x+1)]}{(x-1)^2}$

$f'(x) = \frac{(x+1)(2x - 2 - x - 1)}{(x-1)^2}$

$f'(x) = \frac{(x+1)(x-3)}{(x-1)^2}$

**3. Set the Derivative Equal to Zero:**

To find critical points:

$f'(x) = 0 \Rightarrow \frac{(x+1)(x-3)}{(x-1)^2} = 0$

Since the denominator is never zero (as $x \ne 1$), we set the numerator equal to zero:

$(x+1)(x-3) = 0$

Thus, the critical points are $x = -1$ and $x = 3$.

**4. Determine the Nature of the Extremum (First Derivative Test):**

* **For $x = -1$:**

    * Choose $x = -2$: $f'(-2) = \frac{(-2+1)(-2-3)}{(-2-1)^2} = \frac{(-1)(-5)}{9} = \frac{5}{9} > 0$
    * Choose $x = 0$: $f'(0) = \frac{(0+1)(0-3)}{(0-1)^2} = \frac{(1)(-3)}{1} = -3 < 0$

The derivative changes from positive to negative, so $x = -1$ is a local maximum.

* **For $x = 3$:**

    * Choose $x = 2$: $f'(2) = \frac{(2+1)(2-3)}{(2-1)^2} = \frac{(3)(-1)}{1} = -3 < 0$
    * Choose $x = 4$: $f'(4) = \frac{(4+1)(4-3)}{(4-1)^2} = \frac{(5)(1)}{9} = \frac{5}{9} > 0$

The derivative changes from negative to positive, so $x = 3$ is a local minimum.

**5. Find the Function Values at Critical Points:**

* At $x = -1$:

$f(-1) = \frac{(-1)^2 + 2(-1) + 1}{-1 - 1} = \frac{1 - 2 + 1}{-2} = \frac{0}{-2} = 0$

* At $x = 3$:

$f(3) = \frac{3^2 + 2(3) + 1}{3 - 1} = \frac{9 + 6 + 1}{2} = \frac{16}{2} = 8$

**Conclusion:**

The function $f(x)$ has a local maximum at $(-1, 0)$.

The function $f(x)$ has a local minimum at $(3, 8)$.
