## 19. Limits of Sequences

1. Calculate:
   - $\displaystyle \lim_{n \to \infty} \frac{n^2 + 3n}{2 n^2 - 2n}$

   - $\displaystyle \lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1}$

2. Prove using the squeeze theorem:
   - $\displaystyle\lim_{n \to \infty} \frac{\sin(n)}{n}$

4. Find the limit of the sequence:
   - $a_n = (1+\frac{1}{n})^n$

<h2> SOLUTIONS </h2>
<hr>

# Solution 1

<h3> First equation: </h3>

$\lim_{n\to\infty} \frac{n^2 + 3n}{2n^2 - 2n}$

$= \lim_{n\to\infty} \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}  \quad \text{(Dividing by } n^2)$

$= \frac{1 + 0}{2 - 0}$

$= \frac{1}{2}$

<h3> Second equation: </h3>

$\lim_{n\to\infty} \frac{(2n + 3)^3}{n^3 - 1}$

$= \lim_{n\to\infty} \frac{8n^3 + 36n^2 + 54n + 27}{n^3 - 1}$

$= \lim_{n\to\infty} \frac{8 + \frac{36}{n} + \frac{54}{n^2} + \frac{27}{n^3}}{1 - \frac{1}{n^3}} \quad \text{(Dividing by } n^3)$

$= \frac{8 + 0 + 0 + 0}{1 - 0}$

$= 8$

# Solution 2

**1. Understand the Squeeze Theorem:**

The Squeeze Theorem states that if $f(n) \le g(n) \le h(n)$ for all $n$ (except possibly at $n=a$), and

$\lim_{n\to a} f(n) = \lim_{n\to a} h(n) = L$,

then

$\lim_{n\to a} g(n) = L$.

**2. Apply to $\frac{\sin(n)}{n}$:**

We know that the sine function oscillates between -1 and 1 for all real numbers. Therefore:

$-1 \le \sin(n) \le 1$.

Dividing all parts of the inequality by $n$ (assuming $n > 0$):

$-\frac{1}{n} \le \frac{\sin(n)}{n} \le \frac{1}{n}$.

**3. Evaluate the Limits of the Bounds:**

As $n \to \infty$, both $-\frac{1}{n}$ and $\frac{1}{n}$ approach 0:

$\lim_{n\to\infty} \left(-\frac{1}{n}\right) = 0$ and $\lim_{n\to\infty} \frac{1}{n} = 0$.

**4. Conclude Using the Squeeze Theorem:**

Since $\frac{\sin(n)}{n}$ is "squeezed" between $-\frac{1}{n}$ and $\frac{1}{n}$, and both of these approach 0 as $n \to \infty$, it follows that:

$\lim_{n\to\infty} \frac{\sin(n)}{n} = 0$.

**Final Answer:**

0

# Solution 3

To find the limit of the sequence $a_n = \left(1 + \frac{1}{n}\right)^n$ as $n$ approaches infinity, we can recognize that this is a well-known limit in calculus.

**Step-by-Step Solution:**

1. **Express the Sequence:**

$a_n = \left(1 + \frac{1}{n}\right)^n$

2. **Take the Natural Logarithm:**

To simplify the expression, take the natural logarithm of both sides:

$\ln(a_n) = n \cdot \ln\left(1 + \frac{1}{n}\right)$

3. **Use a Substitution:**

Let $x = \frac{1}{n}$. As $n \to \infty$, $x \to 0$. Rewrite the expression:

$\ln(a_n) = \frac{\ln(1 + x)}{x}$

4. **Apply the Limit:**

We know from calculus that:

$\lim_{x \to 0} \frac{\ln(1 + x)}{x} = 1$

Therefore:

$\lim_{n \to \infty} \ln(a_n) = 1$

5. **Exponentiate to Find the Original Limit:**

Since $\ln(a_n) \to 1$, then:

$\lim_{n \to \infty} a_n = e^1 = e$

**Final Answer:**

$e$
