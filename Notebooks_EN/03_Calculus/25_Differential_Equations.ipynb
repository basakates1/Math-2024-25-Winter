## 25. Differential Equations

1. Solve the following first-order ordinary differential equations:
   - $y'(x)= y(x)$
   - $y'(x) = \frac{1}{2y(x)}$
  
3. Solve the first-order ordinary differential equations using the method of separation of variables for $x$ and $y=y(x)$:

   - $\frac{dy}{dx} = \frac{x}{y}$
   - $\frac{dy}{dx} = \frac{y}{x}$
   - $\frac{dy}{dx} = xy$

4. Solve the second-order ordinary differential equations:

   * $y''(x) + y'(x) = 0$, with boundary conditions $y(0) = 2$ and $y'(0) = -1$

   * $y''(x) - y(x)= 0$, with boundary conditions $y(0) = 2$ and $y'(0) = 0$

   * $\frac{d^2 y(x)}{dx^2} = -\omega^2 y(x)$.
  
5. Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}$:

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

# Solution 1

**Solve $y'(x) = y(x)$:**

This is a separable differential equation. We can rewrite it as:

$\frac{dy}{dx} = y$

Separate the variables:

$\frac{dy}{y} = dx$

Integrate both sides:

$\int \frac{dy}{y} = \int dx$

$\ln|y| = x + C$

Exponentiate both sides to solve for $y$:

$y(x) = e^{x+C} = e^C e^x$

Let $C_1 = e^C$ (where $C_1 > 0$):

$y(x) = C_1 e^x$

Final Answer:

$y(x) = C_1 e^x$

---

**Solve $y'(x) = \frac{1}{2}y(x)$:**

This is also a separable differential equation. Rewrite it as:

$\frac{dy}{dx} = \frac{1}{2}y$

Separate the variables:

$\frac{2}{y} dy = dx$

Integrate both sides:

$\int \frac{2}{y} dy = \int dx$

$2\ln|y| = x + C$

$\ln|y|^2 = x+C$

$|y|^2 = e^{x+C} = e^C e^x$

$y^2 = Ae^x$ where $A = e^C > 0$


Solve for $y$:

$y(x) = \pm \sqrt{Ae^x} = \pm \sqrt{A} e^{x/2}$

Let $C_1 = \pm\sqrt{A}$, then

$y(x) = C_1 e^{x/2}$ where $C_1 \ne 0$.

If $y=0$, then $y'=0$ and $y = \frac{1}{2}y$ is satisfied.  So $y(x) = 0$ is also a solution.

Final Answer:

$y(x) = C_1 e^{x/2}$ where $C_1$ is a real number.

# Solution 2

1. $\frac{dy}{dx} = \frac{y}{x}$

Step 1: Separate the variables $x$ and $y$:

$y \, dy = x \, dx$

Step 2: Integrate both sides:

$\int y \, dy = \int x \, dx$

$\frac{y^2}{2} = \frac{x^2}{2} + C$

where $C$ is the constant of integration.

Step 3: Solve for $y$:

$y^2 = x^2 + 2C$

$y = \pm \sqrt{x^2 + 2C}$

Final Solution:

$y = \pm \sqrt{x^2 + C'}$

where $C' = 2C$ is a new constant.

---

2. $\frac{dy}{dx} = \frac{x}{y}$

Step 1: Separate the variables $x$ and $y$:

$\frac{dy}{y} = \frac{dx}{x}$

Step 2: Integrate both sides:

$\int \frac{dy}{y} = \int \frac{dx}{x}$

$\ln|y| = \ln|x| + C$

where $C$ is the constant of integration.

Step 3: Solve for $y$:

$|y| = e^{\ln|x| + C} = e^{\ln|x|} \cdot e^C = |x| \cdot e^C$

$y = \pm e^C x$

Final Solution:

$y = C'x$

where $C' = \pm e^C$ is a new constant.

---

3. $\frac{dy}{dx} = xy$

Step 1: Separate the variables $x$ and $y$:

$\frac{dy}{y} = x \, dx$

Step 2: Integrate both sides:

$\int \frac{dy}{y} = \int x \, dx$

$\ln|y| = \frac{x^2}{2} + C$

where $C$ is the constant of integration.

Step 3: Solve for $y$:

$|y| = e^{\frac{x^2}{2} + C} = e^{\frac{x^2}{2}} \cdot e^C$

$y = \pm e^C e^{\frac{x^2}{2}}$

Final Solution:

$y = C' e^{\frac{x^2}{2}}$

where $C' = \pm e^C$ is a new constant.

---

Summary of Solutions:

$\frac{dy}{dx} = \frac{y}{x} \rightarrow y = \pm \sqrt{x^2 + C'}$

$\frac{dy}{dx} = \frac{x}{y} \rightarrow y = C'x$

$\frac{dy}{dx} = xy \rightarrow y = C' e^{\frac{x^2}{2}}$

# Solution 3

**Problem 1:** $y''(x) + y'(x) = 0$ with boundary conditions $y(0) = 2$ and $y'(0) = -1$

**Step 1:** Find the characteristic equation.

The differential equation is:

$y''(x) + y'(x) = 0$

Assume a solution of the form $y(x) = e^{rx}$. Substituting into the equation gives:

$r^2e^{rx} + re^{rx} = 0 \implies r^2 + r = 0$

The characteristic equation is:

$r^2 + r = 0$

**Step 2:** Solve the characteristic equation.

$r(r+1) = 0 \implies r = 0$ or $r = -1$

**Step 3:** Write the general solution.

The general solution is:

$y(x) = C_1e^{0x} + C_2e^{-x} = C_1 + C_2e^{-x}$

**Step 4:** Apply the boundary conditions to find $C_1$ and $C_2$.

$y(0) = 2$:

$C_1 + C_2e^0 = C_1 + C_2 = 2$

$y'(x) = -C_2e^{-x}$, so $y'(0) = -1$:

$-C_2e^0 = -C_2 = -1 \implies C_2 = 1$

Substitute $C_2 = 1$ into the first equation:

$C_1 + 1 = 2 \implies C_1 = 1$

**Step 5:** Write the final solution.

$y(x) = 1 + e^{-x}$

---

**Problem 2:** $y''(x) - y(x) = 0$ with boundary conditions $y(0) = 2$ and $y'(0) = 0$

**Step 1:** Find the characteristic equation.

The differential equation is:

$y''(x) - y(x) = 0$

Assume a solution of the form $y(x) = e^{rx}$. Substituting into the equation gives:

$r^2e^{rx} - e^{rx} = 0 \implies r^2 - 1 = 0$

The characteristic equation is:

$r^2 - 1 = 0$

**Step 2:** Solve the characteristic equation.

$r^2 - 1 = 0 \implies r = \pm 1$

**Step 3:** Write the general solution.

The general solution is:

$y(x) = C_1e^x + C_2e^{-x}$

**Step 4:** Apply the boundary conditions to find $C_1$ and $C_2$.

$y(0) = 2$:

$C_1e^0 + C_2e^0 = C_1 + C_2 = 2$

$y'(x) = C_1e^x - C_2e^{-x}$, so $y'(0) = 0$:

$C_1e^0 - C_2e^0 = C_1 - C_2 = 0 \implies C_1 = C_2$

Substitute $C_1 = C_2$ into the first equation:

$C_1 + C_1 = 2 \implies 2C_1 = 2 \implies C_1 = 1$

Thus, $C_2 = 1$.

**Step 5:** Write the final solution.

$y(x) = e^x + e^{-x}$  or $y(x) = 2\cosh(x)$

---

**Problem 3:** $\frac{d^2y(x)}{dx^2} = -\omega^2 y(x)$

**Step 1:** Rewrite the differential equation.

The differential equation is:

$\frac{d^2y(x)}{dx^2} + \omega^2 y(x) = 0$

**Step 2:** Find the characteristic equation.

Assume a solution of the form $y(x) = e^{rx}$. Substituting into the equation gives:

$r^2e^{rx} + \omega^2e^{rx} = 0 \implies r^2 + \omega^2 = 0$

The characteristic equation is:

$r^2 + \omega^2 = 0$

**Step 3:** Solve the characteristic equation.

$r^2 + \omega^2 = 0 \implies r = \pm i\omega$

**Step 4:** Write the general solution.

The general solution is:

$y(x) = C_1\cos(\omega x) + C_2\sin(\omega x)$

**Step 5:** Apply boundary conditions (if provided).

No boundary conditions are given, so the general solution is:

$y(x) = C_1\cos(\omega x) + C_2\sin(\omega x)$

# Solution 4

To determine if the function $\psi(t,x) = A\cos(\omega t + kx)$ is a solution to the wave equation:

$$\frac{\partial^2\psi(t,x)}{\partial t^2} - v^2\frac{\partial^2\psi(t,x)}{\partial x^2} = 0,$$

we need to compute the second partial derivatives of $\psi(t,x)$ with respect to $t$ and $x$, and then substitute them into the wave equation.

**Step 1: Compute $\frac{\partial^2\psi(t,x)}{\partial t^2}$**

First, compute the first partial derivative of $\psi(t,x)$ with respect to $t$:

$$\frac{\partial\psi(t,x)}{\partial t} = -A\omega\sin(\omega t + kx).$$

Next, compute the second partial derivative with respect to $t$:

$$\frac{\partial^2\psi(t,x)}{\partial t^2} = -A\omega^2\cos(\omega t + kx).$$

**Step 2: Compute $\frac{\partial^2\psi(t,x)}{\partial x^2}$**

First, compute the first partial derivative of $\psi(t,x)$ with respect to $x$:

$$\frac{\partial\psi(t,x)}{\partial x} = -Ak\sin(\omega t + kx).$$

Next, compute the second partial derivative with respect to $x$:

$$\frac{\partial^2\psi(t,x)}{\partial x^2} = -Ak^2\cos(\omega t + kx).$$

**Step 3: Substitute into the wave equation**

Substitute the second derivatives into the wave equation:

$$\frac{\partial^2\psi(t,x)}{\partial t^2} - v^2\frac{\partial^2\psi(t,x)}{\partial x^2} = 0.$$

Substituting the expressions we found:

$$-A\omega^2\cos(\omega t + kx) - v^2(-Ak^2\cos(\omega t + kx)) = 0.$$

Simplify the equation:

$$-A\omega^2\cos(\omega t + kx) + Av^2k^2\cos(\omega t + kx) = 0.$$

Factor out $A\cos(\omega t + kx)$:

$$A\cos(\omega t + kx)(-\omega^2 + v^2k^2) = 0.$$

For this equation to hold for all $t$ and $x$, the coefficient must be zero:

$$-\omega^2 + v^2k^2 = 0.$$

This implies:

$$v^2 = \frac{\omega^2}{k^2}.$$

Given that $v = \frac{\omega}{k}$, this condition is satisfied:

$$v^2 = \left(\frac{\omega}{k}\right)^2 = \frac{\omega^2}{k^2}.$$

**Conclusion**

Since the condition $v^2 = \frac{\omega^2}{k^2}$ holds true, the function $\psi(t,x) = A\cos(\omega t + kx)$ is indeed a solution to the wave equation.

Yes, $\psi(t,x) = A\cos(\omega t + kx)$ is a solution to the wave equation.
